  <h2>مقدمه: Collections در جاوا</h2>
  <ul>
  <li>
  Collections در جاوا چارچوبی است که امکان ذخیره سازی و دستکاری مجموعه ای از اشیا را فراهم می کند.
  
  </li>
  <li>
  همه عملیاتی که می توانید برروی داده ای انجام دهید: مانند جستجو (searching) ، مرتب سازی (sorting) ، دستکاری، اضافه نمودن داده جدید، حذف و غیره با استفاده از این چارچوب قابل انجام می باشد.

  </li>
  <li>
  Collection  جاوا واسطهای (interface) بسیاری  نظیر (Set, List, Queue, Deque ,…) و همچنین کلاسهایی نظیر (ArrayList, Vector, LinkedList, PriorityQueue, HashSet, LinkedHashSet, TreeSet etc)  را در اختیار برنامه نویس گذاشته است.
  
  </li>
  <li>
  با collections  می توانید داده ها را از یک ساختار ذخیره سازی (مانند ArrayList) به یک ساختار دیگر(مانند Array) انتقال دهید.
  
  </li>
  </ul>

 <h4 style="color: #dc2626;">Collection چیست؟</h4>
  <ul>
    <li>یک Collection (container) شیئ است که حاوی گروهی از اشیا بوده و به این گروه در قالب یک واحد نگریسته می شود.
    </li>
    <li>انواع مختلفی از اشیا می توانند به عنوان عناصر collections ذخیره سازی، بازیابی و دستکاری شوند. 
    </li>
  </ul>

  <h4 style="color: #dc2626;">مزایای Collections در جاوا</h4>
  <ol>
    <li>کلاس های Collection مانند ArrayList، HashSet، HashMap و ...، امکان ذخیره، جستجو، حذف و مدیریت داده‌ها را بسیار آسان‌تر از آرایه‌ها فراهم می‌کنند.</li>
    <li>در مقایسه با آرایه‌ها که اندازه‌ی ثابتی دارند، بسیاری از کلاس‌های Collection مانند ArrayList یا HashSet به‌صورت پویا (Dynamic) رشد یا کاهش پیدا می‌کنند.</li>
    <li>کلاس‌های Collection شامل متدهای مفیدی مثل ()sort(), reverse(), contains(), addAll(), removeAll و ... هستند که برنامه‌نویس را از نوشتن کدهای تکراری بی‌نیاز می‌کنند.</li>
    <li>استفاده از Collectionها باعث می‌شود کد شما منظم‌تر، خواناتر و آسان‌تر برای نگه‌داری و توسعه باشد.</li>
  </ol>

  <h2>چارچوب Collections</h2>

  <ul>
    <li>یک collection  را شیئی تصور کنید که ریموت کنترلهایی به عناصر (اشیایی دیگر) را نگهداری می کند.
    </li>
    <li>چارچوب collections جاوا بخشی از پکیج <code>java.util</code> را تشکیل می دهد.
    </li>
    <li>یک چارچوب collections از سه بخش اصلی تشکیل شده است:
    <ul>
    <li>
    <strong>Interfaces</strong>: هر واسط، عملیات و قواعدی برای یک نوع   collection مشخص (مانند Queue، Set، List ...) تعریف می کند. 
    </li>
    <p>با دانستن متدهای واسط یک collection از عملکرد و خصوصیات آن آگاه می شوید.
    </p>
    <li><strong>Implementations</strong>: شامل کلاسهایی است که واسطهای بالا را پیاده سازی کرده اند (برای مثال LinkedList، HashSet،...)
    </li>
    <li><strong>Algorithms</strong>: متدهای پلی مورفیک سودمند برای ایجاد و دستکاری اشیایی از کلاسهای بالا مانند Sorting، index searching، replacing ، reversing و غیره
    </li>
    </ul>
    </li>
  </ul>
  <h4 style="color: #dc2626;">سلسله مراتب چارچوب Collections</h4>

  <p>
  پکیج <code>java.util</code> حاوی همه کلاسها و interface های لازم برای چارچوب collection می باشد. 
  
  </p>

 <h4 style="color: #dc2626;">Methods of Collection interface</h4>

  <ul>
  <code>public boolean addAll(collection c)
  </code> 
    <li>is used to insert the specified collection elements in the invoking collection.
    </li>
  <code>public boolean remove(Object element)
  </code>
    <li>is used to delete an element from this collection.
    </li>
  <code>public boolean removeAll(Collection c)
  </code>
    <li>is used to delete all the elements of specified collection from the invoking collection.
    </li>
   <code>public boolean retainAll(Collection c)
  </code>
    <li>
    is used to delete all the elements of invoking collection except the specified collection.
    </li>
     <code>
   public int size()
  </code>
    <li>
    return the total number of elements in the collection.
    </li>
         <code>
  public void clear()
  </code>
    <li>
    removes the total no of element from the collection.
  </li>
  <code>public boolean contains(object element)
  </code> 
    <li>is used to search an element.
    </li>
  <code>public boolean containsAll(Collection c)
  </code>
    <li>is used to search the specified collection in this collection.
    </li>
  <code>
  public Iterator iterator()
  </code>
    <li>
     returns an iterator.
  </li>
  <code>public Object[] toArray()
  </code> 
    <li>converts collection into array.
    </li>
  <code>public boolean isEmpty()
  </code>
    <li>checks if collection is empty.
    </li>
      <code>public boolean equals(Object element)
  </code>
    <li>
     matches two collection.
  </li>
  <code>public int hashCode()
  </code> 
    <li>returns the hashcode number for collection.
    </li>
  </ul>

  <h2>Iterator Interface</h2>
  <ul>
  <li>تا قبل از جاوا 5 ، امکان for each برای پیمایش وجود نداشت.</li>
  <li>قبل از جاوا 5 با کمک iterator پیمایش روی collectionها انجام می‌شد.
  <ul>
  <li>این امکان همچنان وجود دارد و کاربردهایی نیز دارد.</li>
  </ul>
  </li>
  </ul>

<h4 style="color: #dc2626;">سه متد اصلی را تعریف کرده است:</h4>
<ul>
  <li><code>Object next()</code> : عنصر بعدی را در  Collection برمی‌گرداند.</li>
  <li><code>boolean hasNext()</code> : بررسی می‌کند آیا عنصر دیگری برای پیمایش وجود دارد یا خیر.</li>
  <li><code>void remove()</code> : آخرین عنصری را که توسط <code>next()</code> بازگردانده شده، از مجموعه حذف می‌کند.</li>
</ul>


  <ul>
    <li>این سه متد امکان دسترسی به عناصر ذخیره شده در collections را فراهم می کنند:
    </li>
    <li>یک Iterator محل عناصر را در یک Collections می داند.
    </li>
    <li>هربار فراخوانی <code>next()</code>  سبب می شود عنصر بعدی در collection خوانده شود.
    </li>
    <p>می توانید این عنصر را استفاده یا حذف کنید.</p>
  </ul>

  <div class = "images">
    <img src="img/pic3.png"/>
  </div>

  <ul>
    <li>یک Iterator  شیئی است که به شما امکان می دهد یک collection را پیمایش کنید و با قرار گرفتن روی عنصر ذخیره شده در آن collection، در صورت تمایل عنصر را حذف کنید.
</li>
    <li>برای ایجاد شیئی از Iterator برای پیمایش عناصر یک collection، کافیست  متد <code>iterator()</code>  را برروی شیئ collection، فراخوانی کنید.
    
  </li>
    <li>واسط Iterator به صورت زیر تعریف شده است:
    
  </li>
  </ul>

</div>

```java
public interface Iterator<E> {
    boolean hasNext();
    E next();
    void remove(); //optional
}
```

    <h4 style="color: #dc2626;">واسط list در جاوا</h4>
    <p>جاوا واسطی به نام <code>java.util.List</code> دارد که بیانگر لیستی از اشیا می باشد. این واسط، متدهای زیر را به متدهای موجود در Collection اضافه می کند:
  </p>
  <ul>
  <code>public void add(int index, Object o)
  </code> 
    <li>در مکان مشخص شده عنصر مشخصی را به لیست اضافه می کند.
    </li>
  <code>public Object get(int index)
  </code>
    <li>عنصری را در مکان مشخص شده در لیست برمی گرداند.

  </code>
    <li>اندیس اولین رخداد یک عنصر مشخص را در لیست برمی گرداند و در صورت عدم وجود عنصر، مقدار 1- را بر می گرداند.
    </li>
   <code>
   public int lastIndexOf(Object o)
  </code>
    <li>
    اندیس آخرین رخداد یک عنصر مشخص را در لیست برمی گرداند و در صورت عدم وجود عنصر، مقدار 1- را بر می گرداند.
    </li>
     <code>
   public Object remove(int index)
  </code>
    <li>
    عنصری را در مکان مشخص شده در لیست حذف می کند.
    </li>
         <code>
  public Object set(int index, Object o)
  
  </code>
    <li>
    عنصری را در مکان مشخص شده در لیست را با عنصر مشخص شده تعویض می کند.
    
  </li>
  </ul>
   <h4 style="color: #dc2626;">مثال</h4>
</div>

```java
import java.util.ArrayList;

public class MyArrayList {

    public static void main(String args[ ]) {

        ArrayList alist = new ArrayList( );

        alist.add(new String("One"));
        alist.add(new String("Two"));
        alist.add(new String("Three"));

        System.out.println(alist.get(0));
        System.out.println(alist.get(1));
        System.out.println(alist.get(2));
    }
}
```

    </li>
    <li>این کلاس utility، واسط Collection را پیاده سازی کرده است. 
  </li>
  </ul>

  <ul>
  <code>public static void copy(List dest, List src)
  </code> 
    <li>متد copy در کلاس Collections محتوای یک لیست را درلیستی دیگر کپی می کند. بنابراین این کلاس مستقل از اشیای ساخته شده از collection  می باشد. 
    </li>
  <code>public static void fill(List list, Object value)
  </code>
    <li>عنصری را در مکان مشخص شده در لیست برمی گرداند.
    </li>
  <code>public static Object max(Collection coll)
  </code>
    <li>اندیس اولین رخداد یک عنصر مشخص را در لیست برمی گرداند و در صورت عدم وجود عنصر، مقدار 1- را بر می گرداند.
    </li>
   <code>
   public static Object min(Collection coll)
  </code>
    <li>
    اندیس آخرین رخداد یک عنصر مشخص را در لیست برمی گرداند و در صورت عدم وجود عنصر، مقدار 1- را بر می گرداند.
    </li>
     <code>
   public static void reverse(List list)
  </code>
    <li>
    عنصری را در مکان مشخص شده در لیست حذف می کند.
    </li>
         <code>
  public static void shuffle(List list)
  </code>
    <li>
    عنصری را در مکان مشخص شده در لیست را با عنصر مشخص شده تعویض می کند.
  </li>

```java
import java.util.*;
public class LoopExample {

    public static void main(String[] args) {

        ArrayList<Integer> arrlist = new ArrayList<Integer>();
        arrlist.add(14);
        arrlist.add(7);
        arrlist.add(39);
        arrlist.add(40);

        /* For Loop for iterating ArrayList */
        System.out.println("For Loop");
        for (int counter = 0; counter < arrlist.size(); counter++) {
            System.out.println(arrlist.get(counter));
        }
        /* Advanced For Loop */
        System.out.println("Advanced For Loop");
        for (Integer num : arrlist) {
            System.out.println(num);
        }

        /* While Loop for iterating ArrayList */
        System.out.println("While Loop");
        int count = 0;
        while (arrlist.size() > count) {
            System.out.println(arrlist.get(count));
            count++;
        }

        /* Looping ArrayList using Iterator */
        System.out.println("Iterator");
        Iterator iter = arrlist.iterator();
        while (iter.hasNext()) {
            System.out.println(iter.next());
        }
    }
}
```

  <h2>سنکرون سازی ArrayList</h2>
  <ul>
  <li>ArrayList آسنکرون بوده و درمحیطی چندتریدی می تواند  همزمان توسط چندین ترید مورد دستیابی و تغییر قرار گیرد.
  </li>
  <li>اگر نخواهیم در یک لحظه دسترسی به آن توسط چند ترید صورت گیرد، باید آن را سنکرون سازی کنیم تا در یک لحظه میان چندین ترید به اشتراک گذاشته نشود.
  </li>
  <li>سنکرون سازی آشکار، به دو روش صورت می گیرد:
  <ul>
  <li>استفاده از متد <code>Collections.synchronizedList()</code> 
  </li>
  <li>استفاده از نسخه thread-safe لیست : <code>CopyOnWriteArrayList</code>
  </li>
  </ul>
  </li>
  <li>در هنگام استفاده از iterator برروی یک شیئ ArrayList بهتر است آن را در بلاک سنکرون قرار دهیم.
  </li>
  </ul>

  <h4 style="color: #dc2626;">روش اول</h4>

```java
import java.util.ArrayList;
import java.util.Iterator;
import java.util.List;
import java.util.Collections;

public class Details {

    public static void main(String a[]){
        List<String> syncal = Collections.synchronizedList(new ArrayList<String>());

        //Adding elements to synchronized ArrayList
        syncal.add("Pen");
        syncal.add("NoteBook");
        syncal.add("Ink");

        System.out.println("Iterating synchronized ArrayList:");
        synchronized(syncal) {
            Iterator<String> iterator = syncal.iterator();
            while (iterator.hasNext())
                System.out.println(iterator.next());
        }
    }
}
```

  <ul>

    <br>

    <br>
    اما لیست پیوندی doubly linked list  را پیاده سازی کرده که نیاز به پیمایش همه عناصر برای جستجوی یک عنصر را دارد.
    </li>
    <br>

<li><strong>حذف</strong> : عملیات حذف درLinkedList  با درنظر نگرفتن زمان رسیدن به اندیس موردنظر، دارای پیچیدگی O(1) است در حالی که ArrayList  دارای پیچیدگی زمانی O(n)  در بدترین حالت (حذف اولین عنصر) و O(1) در بهترین حالت است (حذف عنصر  آخر)
    <ul>
    <br>
    <li>توجه داشته باشید اگر پیمایش LinkedList جهت رسیدن به اندیس عنصری که می خواهیم آن را حذف کنیم،  در نظر گرفته شود، زمان حذف افزایش خواهد یافت و ممکن است بهبود زمانی LinkedList نسبت به ArrayList  قابل توجه نباشد. 
    </li>
    <li>اما در مواقعی که حذف عناصر از اندیسهای اول LinkedList به طور متوالی انجام شود، به دلیل زمان پیمایش کمتر و عدم نیاز به عملیات شیفت سرعت LinkedList به مراتب از بیشتر از ArrayList خواهد شد. 
    </li>
    </ul>
    <strong>نتیجه گیری</strong> : حذف عنصر LinkedList در حالت متوسط (اغلب اوقات) سریعتر از ArrayList می باشد.

```java
import java.util.*;

public class VectorExample {

    public static void main(String args[]) {
        /* Vector of initial capacity(size) of 2 */
        Vector<String> vec = new Vector<String>(2);
        /* Adding elements to a vector*/
        vec.addElement("Apple");
        vec.addElement("Orange");
        vec.addElement("Mango");
        vec.addElement("Fig");
        /* check size and capacityIncrement */
        System.out.println("Size is: " + vec.size());
        System.out.println("Default capacity increment is: " + vec.capacity());
        vec.addElement("fruit1");
        vec.addElement("fruit2");
        vec.addElement("fruit3");

        /* size and capacityIncrement after two insertions */
        System.out.println("Size after addition: " + vec.size());
        System.out.println("Capacity after increment is: " + vec.capacity());

        /* Display Vector elements */
        Enumeration en = vec.elements();
        System.out.println("\nElements are:");
        while (en.hasMoreElements())
            System.out.print(en.nextElement() + " ");
    }
}
```
