<div class="card" dir="rtl">
  <h2 style="margin-bottom: 4px;">اعلان / ایجاد آرایه‌های دوبعدی  </h2>

  <p>
    آرایه‌های دوبعدی برای ذخیره‌سازی داده‌ها به صورت جدول یا ماتریس استفاده می‌شوند. در ادامه روش‌های مختلف اعلان و ایجاد آن‌ها را بررسی می‌کنیم.
  </p>

  <div style="background-color:#f0f8ff; padding:10px; border-radius:8px; margin-bottom:10px;">
    <strong> ۱. اعلان آرایه دوبعدی:</strong>
    <p>در این روش فقط نوع و نام آرایه مشخص می‌شود، بدون تخصیص حافظه.</p>
    <pre style="background-color:#272822; color:#f8f8f2; font-weight:bold; padding:12px; border-radius:8px; direction:ltr; text-align:left;">
dataType[][] array_name;
    </pre>
  </div>

  <div style="background-color:#eaffea; padding:10px; border-radius:8px; margin-bottom:10px;">
    <strong> ۲. ایجاد آرایه در heap و انتساب به متغیر:</strong>
    <p>در این مرحله حافظه اختصاص داده می‌شود و آرایه آماده استفاده است.</p>
    <pre style="background-color:#272822; color:#f8f8f2; font-weight:bold; padding:12px; border-radius:8px; direction:ltr; text-align:left;">
array_name = new dataType[10][10];
    </pre>
  </div>

  <div style="background-color:#fff8dc; padding:10px; border-radius:8px; margin-bottom:10px;">
    <strong> ۳. ترکیب اعلان و ایجاد:</strong>
    <p>این روش رایج‌ترین و توصیه‌شده‌ترین روش در جاوا است.</p>
    <pre style="background-color:#272822; color:#f8f8f2; font-weight:bold; padding:12px; border-radius:8px; direction:ltr; text-align:left;">
dataType[][] array_name = new dataType[10][10];
    </pre>
  </div>

  <div style="background-color:#ffecec; padding:10px; border-radius:8px;">
    <strong style="color:red;"> ۴. روش جایگزین (سبک قدیمی‌تر):</strong>
    <p>در این روش جایگاه براکت‌ها متفاوت است.</p>
    <pre style="background-color:#272822; color:#f8f8f2; font-weight:bold; padding:12px; border-radius:8px; direction:ltr; text-align:left;">
dataType array_name[][] = new dataType[10][10];
    </pre>
  </div>

  <p style="margin-top:15px;">
     در جاوا، آرایه‌های دوبعدی در واقع آرایه‌ای از آرایه‌ها هستند. یعنی هر سطر یک آرایه مستقل است. این ساختار امکان ایجاد آرایه‌های نامتقارن <em>(jagged arrays)</em> را نیز فراهم می‌کند.
  </p>

  <h2 style="margin-bottom: 4px;">اعلان و مقداردهی آرایه‌های دوبعدی</h2>


  <div style="background-color:#f0f8ff; padding:10px; border-radius:8px; margin-bottom:10px;">
    <strong>۱. اعلان و ایجاد آرایه:</strong>
    <p>دو سبک نحوی برای ایجاد آرایه وجود دارد:</p>
    <pre style="background-color:#272822; color:#f8f8f2; font-weight:bold; padding:12px; border-radius:8px; direction:ltr; text-align:left;">
int[][] matrix = new int[10][10];
// or
int matrix[][] = new int[10][10];
    </pre>
  </div>

  <div style="background-color:#eaffea; padding:10px; border-radius:8px; margin-bottom:10px;">
    <strong>۲. مقداردهی دستی:</strong>
    <p>می‌توان به‌صورت مستقیم به عناصر آرایه مقدار داد:</p>
    <pre style="background-color:#272822; color:#f8f8f2; font-weight:bold; padding:12px; border-radius:8px; direction:ltr; text-align:left;">
matrix[0][0] = 3;
    </pre>
    <p style="color:#555; margin-top:6px;">
      این دستور مقدار ۳ را در خانه‌ی سطر صفرم و ستون صفرم قرار می‌دهد.
    </p>
  </div>

  <div style="background-color:#fff8dc; padding:10px; border-radius:8px; margin-bottom:10px;">
    <strong>۳. مقداردهی با حلقه و مقادیر تصادفی:</strong>
    <p>برای پر کردن آرایه با مقادیر تصادفی از حلقه‌های تو در تو استفاده می‌کنیم:</p>
    <pre style="background-color:#272822; color:#f8f8f2; font-weight:bold; padding:12px; border-radius:8px; direction:ltr; text-align:left;">
for (int i = 0; i < matrix.length; i++)
    for (int j = 0; j < matrix[i].length; j++)
        matrix[i][j] = (int)(Math.random() * 1000);
    </pre>
    <p style="color:#555; margin-top:6px;">
      تابع <code>()Math.random</code> عددی بین ۰ تا ۱ تولید می‌کند که با ضرب در ۱۰۰۰ و تبدیل به عدد صحیح، مقدار نهایی را می‌سازد.
    </p>
  </div>


  <div style="background-color:#f9f9f9; padding:10px; border-radius:8px; margin-top:10px;">
    <strong>اعلان آرایه‌های اعشاری:</strong>
    <p>می‌توان آرایه‌هایی با نوع داده‌ی <code>double</code> نیز تعریف کرد:</p>
    <pre style="background-color:#272822; color:#f8f8f2; font-weight:bold; padding:12px; border-radius:8px; direction:ltr; text-align:left;">
double[][] x;
    </pre>
  </div>

  <div style="background-color:#ffecec; padding:10px; border-radius:8px;">
    <strong>نکته مهم:</strong>
    <p>
      هنگام استفاده از حلقه‌ها، از <code>matrix.length</code> برای تعداد سطرها و <code>matrix[i].length</code> برای تعداد ستون‌ها استفاده کنید تا از خطای <em>ArrayIndexOutOfBounds</em> جلوگیری شود.
    </p>
  </div>
</div>

  <h2 style="margin-bottom: 4px;">جمع دو ماتریس</h2>

  <div style="background-color:#f0f8ff; padding:10px; border-radius:8px; margin-bottom:12px;">
    <strong>صورت مسئله:</strong>
    <p style="margin-top:6px;">
      دو ماتریس هم‌ابعاد داریم (۲×۳). هدف این است که ماتریس حاصل جمع را با جمع عنصر به عنصر
      به‌دست بیاوریم؛ یعنی هر خانه از ماتریس خروجی برابر مجموع مقدار متناظر در ماتریس‌های ورودی باشد.
      در نهایت، مقادیر ماتریس جمع به‌صورت سطر به سطر چاپ شوند.
    </p>
  </div>

```java
class MatrixAddition {
    public static void main(String[] args) {
        // define two input matrices (2 rows, 3 columns)
        int[][] a = { {1, 3, 4}, {3, 4, 5} };
        int[][] b = { {1, 3, 4}, {3, 4, 5} };

        // result matrix to store element-wise sum
        int[][] c = new int[2][3];

        // compute sum and print row by row
        for (int i = 0; i &lt; 2; i++) {
            for (int j = 0; j &lt; 3; j++) {
                c[i][j] = a[i][j] + b[i][j];
                System.out.print(c[i][j] + " ");
            }
            System.out.println();
        }
    }
}
```

<div class="card" dir="rtl">
  <h2 style="margin-bottom: 4px;">آرایه‌های نامتقارن (Ragged Arrays)</h2>

  <!-- توضیح کلی -->
  <p>
    در زبان‌هایی مانند جاوا، آرایه‌های دوبعدی در واقع <strong>آرایه‌ای از آرایه‌ها</strong> هستند. 
    این یعنی هر سطر یک آرایه مستقل است و می‌تواند طول متفاوتی داشته باشد. 
    به چنین ساختاری <strong>آرایه نامتقارن</strong> گفته می‌شود. 
    این ویژگی باعث انعطاف‌پذیری بیشتر نسبت به آرایه‌های منظم می‌شود و امکان مدل‌سازی داده‌هایی با ابعاد نامنظم را فراهم می‌کند.
  </p>

  <p>
    <strong>ویژگی‌های کلیدی:</strong>
    <ul>
      <li>هر سطر می‌تواند طول متفاوتی داشته باشد.</li>
      <li>حافظه بهینه‌تر مصرف می‌شود، چون فقط به اندازه‌ی موردنیاز هر سطر حافظه تخصیص داده می‌شود.</li>
      <li>برای نمایش ساختارهایی مثل مثلث، داده‌های نامنظم یا ماتریس‌های با طول متغیر بسیار مناسب است.</li>
    </ul>
  </p>

  <!-- مثال اول -->
  <div style="background-color:#f0f8ff; padding:10px; border-radius:8px; margin-bottom:15px;">
    <strong>مثال :</strong>
    <pre style="background-color:#272822; color:#f8f8f2; font-weight:bold;
    padding:12px; border-radius:8px; direction:ltr; text-align:left;">
int[][] matrix = {
  {1, 2, 3, 4, 5},
  {2, 3, 4, 5},
  {3, 4, 5},
  {4, 5},
  {5}
};
matrix.length      // 5
matrix[0].length   // 5
matrix[1].length   // 4
matrix[2].length   // 3
matrix[3].length   // 2
matrix[4].length   // 1
    </pre>
  </div>

  <h2 style="margin-bottom: 4px;">پردازش آرایه‌های چندبعدی</h2>

  <p>

```java
import java.util.Scanner;

public class PassTwoDimensionalArray {
    public static void main(String[] args) {
        Scanner input = new Scanner(System.in);

        // Enter array values
        int[][] m = new int[3][4];
        System.out.println("Enter " + m.length + " rows and " + m[0].length + " columns: ");
        for (int row = 0; row < m.length; row++)
            for (int column = 0; column < m[row].length; column++)
                m[row][column] = input.nextInt();

        // Display result
        System.out.println("\nSum of all elements is " + sum(m));
    }

    public static int sum(int[][] m) {
        int total = 0;
        for (int row = 0; row < m.length; row++)
            for (int column = 0; column < m[row].length; column++)
                total += m[row][column];

        return total;
    }
}
```

  <p style="margin-top:18px;">
    برای یافتن نزدیک‌ترین دو نقطه، می‌توان فاصله بین هر جفت را با فرمول اقلیدسی محاسبه کرد:  
    <span style="direction:ltr; display:inline-block; font-weight:bold;">√((x₂ - x₁)² + (y₂ - y₁)²)</span>  
    سپس کمترین مقدار را انتخاب کرد.  
  </p>
</div>

```java
import java.util.Scanner;

public class FindNearestPoints {
    public static void main(String[] args) {
        Scanner input = new Scanner(System.in);

        // Prompt for number of points
        System.out.print("Enter the number of points: ");
        int numberOfPoints = input.nextInt();

        // Create array to store points
        double[][] points = new double[numberOfPoints][2];
        System.out.println("Enter " + numberOfPoints + " points:");
        for (int i = 0; i < numberOfPoints; i++) {
            points[i][0] = input.nextDouble(); // x-coordinate
            points[i][1] = input.nextDouble(); // y-coordinate
        }

        // Initialize with first two points
        int p1 = 0, p2 = 1;
        double shortestDistance = distance(points[p1][0], points[p1][1],
                                           points[p2][0], points[p2][1]);

        // Compare all pairs to find the closest
        for (int i = 0; i < numberOfPoints; i++) {
            for (int j = i + 1; j < numberOfPoints; j++) {
                double d = distance(points[i][0], points[i][1],
                                    points[j][0], points[j][1]);

                if (d < shortestDistance) {
                    p1 = i;
                    p2 = j;
                    shortestDistance = d;
                }
            }
        }

        // Display result
        System.out.println("The closest two points are: (" +
                           points[p1][0] + ", " + points[p1][1] + ") and (" +
                           points[p2][0] + ", " + points[p2][1] + ")");
    }

    // Method to compute Euclidean distance
    public static double distance(double x1, double y1, double x2, double y2) { 
        return Math.sqrt((x2 - x1) * (x2 - x1) +                    
                         (y2 - y1) * (y2 - y1));
    }
}                                               
```

    در برخی مسائل، نیاز داریم داده‌ها را در ساختارهایی پیچیده‌تر از آرایه‌های یک‌بعدی یا دوبعدی ذخیره کنیم. آرایه‌های چندبعدی این امکان را فراهم می‌کنند تا داده‌ها را در قالب ماتریس‌ها، مکعب‌های عددی یا ساختارهای n‌بعدی سازمان‌دهی کنیم.
  </p>

  <p style="margin-top:12px;">
    در زبان جاوا، می‌توان آرایه‌هایی با بیش از دو بعد تعریف کرد. ساختار تعریف آن مشابه آرایه‌های دوبعدی است، با این تفاوت که تعداد براکت‌ها و اندازه‌ها بیشتر می‌شود.
  </p>

    و اندیس سوم <strong>بخش امتحان</strong> را.

  <p style="margin-top:12px;">
    فرض کنید ۷ دانشجو و ۵ امتحان داریم و هر امتحان دو بخش دارد:  
    یک بخش <strong>چندگزینه‌ای</strong> و یک بخش <strong>تشریحی</strong>.
  </p>

  <p style="margin-top:12px;">
    بنابراین <code>scores[i][j][0]</code> بیانگر امتیاز دانشجوی <strong>iام</strong>

  <h2 style="margin-bottom: 4px;">ساختار لیست متغیر آرگومان‌ها  (Varargs) </h2>

  <p>
    ساختار <strong>varargs</strong> (لیست متغیر آرگومان‌ها) به یک متد امکان می‌دهد 
    صفر یا چند آرگومان ورودی بپذیرد.
  </p>

  <p style="margin-top:12px;">
    در صورت عدم استفاده از <strong>varargs</strong> می‌توانیم از متدهای 
    <strong>سربارگذاری‌شده</strong> استفاده کنیم یا یک <strong>آرایه</strong> را به عنوان پارامتر ارسال نماییم.  
    اما این روش‌ها به علت مشکلات نگهداری، مناسب به حساب نمی‌آیند.
  </p>

  <p style="margin-top:12px;">
    در مواقعی که ندانیم چه تعداد آرگومان را به یک متد ارسال کنیم، 
    <strong>varargs</strong> روشی بسیار مناسب خواهد بود.
  </p>

  <p style="margin-top:12px;">
    <strong>مزیت varargs:</strong>
    <ul style="margin-top:8px; line-height:1.8;">
      <li>به متد سربارگذاری‌شده نیاز نداریم.</li>
      <li>میزان کدنویسی کاهش می‌یابد.</li>
      <li>انعطاف‌پذیری بیشتری در ارسال آرگومان‌ها خواهیم داشت.</li>
    </ul>
  </p>

  <p style="margin-top:12px;">
    <strong>قاعده نحوی:</strong>  
    Varargs از سه نقطه پشت سر هم <code>...</code> پس از نوشتن نوع داده استفاده می‌کند:
  </p>

  <div style="background-color:#f9f9f9; padding:12px; border-radius:8px;">
    <pre style="background-color:#272822; color:#f8f8f2; font-weight:bold;
    padding:12px; border-radius:8px; direction:ltr; text-align:left;">
//return_type method_name(data_type...VariableName){}
public static int sum(int... numbers) {
    int total = 0;
    for (int n : numbers) {
        total += n;
    }
    return total;
}
public static void main(String[] args) {
    System.out.println(sum());              // خروجی: 0
    System.out.println(sum(5));             // خروجی: 5

}

  </p>

    <li>در صورت استفاده از چند آرگومان، آرگومان متغیر (<code>varargs</code>) باید آخرین آرگومان باشد.</li>

  <p>

    به عبارتی، <strong>چرخ را دوباره اختراع نکنید!</strong> از ابزارهای آماده‌ی جاوا استفاده کنید.
  </p>

  <!-- ویژگی‌ها -->
  <p>
    <strong>ویژگی‌های کلیدی کلاس Arrays:</strong>
    <ul>
      <li>دارای متدهایی برای <strong>مرتب‌سازی آرایه‌ها</strong> به صورت صعودی یا سفارشی با استفاده از مقایسه‌گرها.</li>

      <li>متدهای کلاس برای انواع داده‌های اولیه (int, double, char...) و اشیاء سربارگذاری شده‌اند.</li>

  <h2 style="margin-bottom: 4px;">کلاس ArrayList و نحوهٔ اعلان آن</h2>

  <!-- توضیح کلی درباره Placeholder -->
  <p>
    کلاس <strong>&lt;ArrayList&lt;T</strong> در جاوا با استفاده از <strong>نوع عمومی (Generic)</strong> تعریف می‌شود.

  </p>

    در مثال زیر، یک لیست از نوع <code>String</code> تعریف شده که می‌تواند رشته‌ها را به‌صورت پویا ذخیره کند:

  <!-- مثال کد -->
  <div style="background-color:#eaffea; padding:10px; border-radius:8px; margin-bottom:15px;">
    <strong>مثال :</strong>

```java
import java.util.Arrays;

public class ArrayManipulations {
    public static void main( String[] args ) {
        // sort doubleArray in ascending order
        double[] doubleArray = { 8.4, 9.3, 0.2, 7.9, 3.4 };
        Arrays.sort( doubleArray );
        System.out.printf( "%ndoubleArray: " );
        for ( double value : doubleArray )
            System.out.printf( "%.1f ", value );

        // fill 10-element array with 7s
        int[] filledIntArray = new int[10];
        Arrays.fill( filledIntArray, 7 );
        displayArray( filledIntArray, "filledIntArray" );

        // copy array a into array b
        int[] intArray = { 1, 2, 3, 4, 5, 6 };
        int[] intArrayCopy = new int[intArray.length];
        System.arraycopy( intArray, 0, intArrayCopy, 0, intArray.length );
        displayArray( intArray, "intArray" );
        displayArray( intArrayCopy, "intArrayCopy" );

        // compare array a and b for equality
        boolean b = Arrays.equals( intArray, intArrayCopy );
        System.out.printf( "%nintArray %s intArrayCopy%n", 
            ( b ? "==" : "!=" ) );
        // compare intArray and filledIntArray for equality
        b = Arrays.equals( intArray, filledIntArray );
        System.out.printf( "%nintArray %s filledIntArray%n", 
            ( b ? "==" : "!=" ) );

        // search intArray for the value 5
        int location = Arrays.binarySearch( intArray, 5 );

        if ( location >= 0 )
            System.out.printf( "Found 5 at element %d in intArray%n", location );
        else
            System.out.printf( "5 not found in intArray%n" );

        // search intArray for the value 8763
        location = Arrays.binarySearch( intArray, 8763 );

        if ( location >= 0 )
            System.out.printf( "Found 8763 at element %d in intArray%n", location );
        else
            System.out.printf( "8763 not found in intArray%n" );
    }

    // output values in each array
    public static void displayArray( int[] array, String description ) {
        System.out.printf( "%n%s: ", description );

        for ( int value : array )
            System.out.printf( "%d ", value );
    }
} // end class ArrayManipulations
```

  <h2 style="margin-bottom: 4px;">مثال ArrayList با عناصر تعریف شده توسط کاربر</h2>

  <!-- توضیح کلی -->
  <p>
    در این مثال، ابتدا کلاس <code>Student</code> تعریف می‌شود که شامل سه ویژگی است: شماره دانش‌آموز، نام و سن. سپس با استفاده از کلاس <code>ArrayList</code>، مجموعه‌ای از دانش‌آموزان ساخته شده و اطلاعات آن‌ها چاپ می‌شود.
  </p>

  <!-- کد برنامه -->
  <div style="background-color:#eaffea; padding:10px; border-radius:8px; margin-bottom:15px;">
    <pre style="background-color:#272822; color:#f8f8f2; font-weight:bold;
    padding:12px; border-radius:8px; direction:ltr; text-align:left;">
import java.util.*;
class Student {
    int rollno;
    String name;
    int age;
    Student(int rollno, String name, int age) {
        this.rollno = rollno;
        this.name = name;
        this.age = age;
    }
}
public class StudentList {
    public static void main(String args[]) {
        //Creating user-defined class objects
        Student s1 = new Student(101, "Sonoo", 23);
        Student s2 = new Student(102, "Ravi", 21);
        Student s3 = new Student(103, "Hanumat", 25);
        ArrayList&lt;Student&gt; list = new ArrayList&lt;Student&gt;();//creating arraylist
        list.add(s1);
        list.add(s2);
        list.add(s3);
        Iterator&lt;Student&gt; itr = list.iterator();
        while (itr.hasNext()) {
            Student st = itr.next();
            System.out.println(st.rollno + " " + st.name + " " + st.age);
        }
    }
}
    </pre>
  </div>

  <!-- توضیحات تخصصی کامل -->
  <div style="background-color:#fff8e1; border-right:4px solid #ffc107; padding:15px; border-radius:8px; line-height:1.9;">
    <strong>  توجه :</strong>
    <p>
       با اینکه اشیای موجود در <code>ArrayList</code> از کلاس <code>Student</code> هستند، چون ریموت کنترل <code>itr</code> از نوع <code>Iterator</code> تعریف شده که یک Interface عمومی است، نمی‌تواند به صفات ویژه شیء <code>Student</code> دسترسی پیدا کند.
    </p>
    <p>
      در نتیجه، ابتدا باید یک <em>downcasting</em> انجام دهیم تا ریموت کنترل بتواند به همه فیلدهای شیء <code>Student</code> دسترسی پیدا کند. این downcasting باید همواره به کلاس واقعی شیء موجود در <code>ArrayList</code> انجام شود تا بتوان از ویژگی‌های اختصاصی آن استفاده کرد.
    </p>
    <p>
      نکته مهم دیگر این است که اگر در اعلان یک <code>ArrayList</code> از placeholder یا همان نوع generic استفاده نشود، به معنای تعریف مجموعه‌ای غیرژنریک یا سنتی خواهد بود. در این حالت، عناصر موجود در <code>ArrayList</code> از نوع کلاس <code>Object</code> خواهند بود.
    </p>
    <p>
      اگر عناصر یک <code>ArrayList</code> از نوع <code>Object</code> باشند، می‌توانند اشیای مختلفی را در حافظه heap کنترل کنند. اما چون ریموت کنترل‌های ذخیره‌شده در <code>ArrayList</code> همگی از نوع <code>Object</code> هستند، برای دسترسی به صفات ویژه اشیای کنترل‌شده توسط آن‌ها، باید حتماً <em>downcasting</em> به کلاس اصلی هر شیء صورت گیرد.
    </p>
  </div>

</div>

```java
import java.util.*;
class Student {
    int rollno;
    String name;
    int age;
    Student(int rollno, String name, int age) {
        this.rollno = rollno;
        this.name = name;
        this.age = age;
    }
}
public class StudentList {
    public static void main(String args[]) {
      //Creating user-defined class objects
        Student s1 = new Student(101, "Sonoo", 23);
        Student s2 = new Student(102, "Ravi", 21);
        Student s3 = new Student(103, "Hanumat", 25);
        ArrayList&lt;Student&gt; list = new ArrayList&lt;Student&gt;();//creating arraylist
        list.add(s1);
        list.add(s2);
        list.add(s3);
        Iterator&lt;Student&gt; itr = list.iterator();
        while (itr.hasNext()) {
            Student st = itr.next();
            System.out.println(st.rollno + " " + st.name + " " + st.age);
        }
    }
}
```
