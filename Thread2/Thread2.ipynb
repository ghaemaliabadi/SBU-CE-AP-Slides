<div class="card" dir="rtl">
  <h2 style="margin-bottom: 20px;">Daemon Thread در جاوا</h2>
  <p> </p>
  <li>
    <strong>Daemon Thread</strong> یک ترد در جاوا می باشد که برای ارائه خدمات به ترد کاربر  (<strong>User Thread</strong>) استفاده می‌شود.
  </li>
  <p style="margin-bottom: 20px; text-indent: 20px;"> 
    درواقع Daemon Thread یک ترد خدماتی (<strong>Service Thread</strong>) است که کارهای پشتیبانی در پس زمینه (<strong>Background Supporting Tasks</strong>) را برای User Thread ها انجام می‌دهد.
  </p>
  <li>
    عمر   Daemon Thread وابسته به ترد های کاربر است.
  </li>
  <p style="margin-bottom: 20px; text-indent: 20px;"> 
  یعنی وقتی همه‌ی User Thread ها تمام شوند، <strong>JVM</strong> به صورت خودکار Daemon Thread ها را متوقف می‌کند.
  </p>
  <li style="margin-top: 20px;">
    به صورت خودکار <strong>Daemon Thread</strong> هایی وجود دارند که در حال اجرا هستند.
  </li>

  <p style="margin-bottom: 3px; text-indent: 20px;">
    مانند Garbage collector،  که برای جمع آوری زباله ها و مدیریت حافظه استفاده می‌شود،
  </p>

  <p style="margin-bottom: 20px; text-indent: 20px;">
    و finalizer که برای آزاد کردن منابع قبل از حذف شئ استفاده می‌شود.
  </p>

  <li style="margin-bottom: 20px;">
    معمولاً یک ترد با اولویت پایین (Low Priority Thread) است.
  </li>

</div>

  <h2 style="margin-bottom: 20px;">  چرا JVM در صورت عدم وجود نخ کاربر، نخ daemon را خاتمه می‌دهد؟ </h2>

  <li style="margin-bottom: 20px;">
    هدف اصلی Daemon Thread ارائه‌ی خدمات به User Thread ‌ها برای انجام کارهای پشتیبان در پس‌زمینه است.
  </li>

  <li style="margin-bottom: 20px;">
    اگر هیچ User Threadی وجود نداشته باشد، نگه داشتن Daemon Thread بی‌معنی است، چون دیگر چیزی برای خدمت‌رسانی وجود ندارد.
  </li>

  <li style="margin-bottom: 20px;">
    به همین دلیل JVM به صورت خودکار Daemon Threadها را خاتمه می‌دهد زمانی که همه‌ی User Threadها تمام شده باشند.
  </li>

<div class="card" dir="rtl">
  <h2 style="margin-bottom: 20px;">  مثال از Daemon thread در جاوا </h2>

  

  <p style="margin-bottom: 20px;"> </p>

</div>

```java
public class TestDaemonThread1 extends Thread {
    public void run() {
        if (Thread.currentThread().isDaemon()) { // checking for daemon thread
            System.out.println("daemon thread work");
        } else {
            System.out.println("user thread work");
        }
    }

    public static void main(String[] args) {
        TestDaemonThread1 t1 = new TestDaemonThread1(); // creating thread
        TestDaemonThread1 t2 = new TestDaemonThread1();
        TestDaemonThread1 t3 = new TestDaemonThread1();

        t1.setDaemon(true); // now t1 is daemon thread

        // starting threads
        t1.start();
        t2.start();
        t3.start();
    }
}

TestDaemonThread1.main(new String[]{})
```

  <h2>ThreadGroup در جاوا</h2>
  
  <li>
    جاوا راهی ساده برای گروه‌ بندی چند <strong>Thread</strong> در یک شیء (<strong>Object</strong>) فراهم کرده است.
  </li>

  <li>
    با استفاده از ThreadGroup می‌توانیم چندین Thread را با یک فراخوانی متد:
  </li>
  <p style="margin-bottom: 4px; text-indent: 35px;"> 
    متوقف (<strong>suspend</strong>)،
  </p>  
  <p style="margin-bottom: 4px; text-indent: 35px;"> 
    از سر بگیریم (<strong>resume</strong>)،
  </p> 
  <p style="margin-bottom: 20px; text-indent: 35px;"> 
    یا قطع (<strong>interrupt</strong>) کنیم.
  </p> 

  <li style="margin-bottom: 30px;">
    در جاوا، گروه ترد ها با کلاس   <strong>java.lang.ThreadGroup</strong>   پیاده سازی شده است.
  </li>

  <li>
    هر <strong>ThreadGroup</strong> مجموعه‌ای از Threadها را نمایش می‌دهد.
  </li>

  <li>
    یک <strong>ThreadGroup</strong> می‌تواند گروه‌های دیگر از ThreadGroupها را هم شامل شود.
  </li>

  <li>
    در نتیجه ساختار آن به شکل درختی (Tree Structure) است:
  </li>

  <p style="margin-bottom: 30px; text-indent: 35px;"> 
    به جز گروه اولیه (initial thread group)، هر گروه والد (parent) دارد.
  </p> 

  <li>
    هر <strong>Thread</strong> فقط می‌تواند به اطلاعات گروه خودش (ThreadGroup خود) دسترسی داشته باشد.
  </li>

  <li style="margin-bottom: 30px;">
    هیچ <strong>Threadی</strong> اجازه ندارد اطلاعات مربوط به گروه والد یا گروه‌های دیگر را بخواند یا تغییر دهد.
  </li>

  <li>
    تنها ۲ <strong>سازنده</strong> برای این کلاس وجود دارند : 
  </li>

<div class="card" dir="rtl">
  <h2>مثال ThreadGroup</h2>
  

  <pre style="background-color: #272822; color: #f8f8f2; padding: 10px; border-radius: 8px; direction: ltr; text-align: left;">
ThreadGroup tg1 = <span style="color: #22e1faff;">new</span> ThreadGroup(<span style="color: #ae6afcff;">"Group A"</span>);
Thread1 = <span style="color: #22e1faff;">new</span> Thread(tg1, <span style="color: #22e1faff;">new</span> MyRunnable(), <span style="color: #ae6afcff;">"one"</span>);
Thread2 = <span style="color: #22e1faff;">new</span> Thread(tg1, <span style="color: #22e1faff;">new</span> MyRunnable(), <span style="color: #ae6afcff;">"two"</span>);
Thread3 = <span style="color: #22e1faff;">new</span> Thread(tg1, <span style="color: #22e1faff;">new</span> MyRunnable(), <span style="color: #ae6afcff;">"three"</span>);
</pre>
  
<li>
در این مثال، سه تا Thread داریم که همگی به یک گروه تعلق دارند.
</li>
<p style="text-indent: 35px;">
نام گروه: Group A
</p>
<p style="text-indent: 35px;">
کلاس: MyRunnable (کلاسی که رابط Runnable را پیاده‌ سازی کرده است)
</p>
<p style="text-indent: 35px;">
نام تردها: one, two, three
</p>

</div>

```java
public class ThreadGroupDemo implements Runnable {
    public void run() {
        System.out.println(Thread.currentThread().getName());
    }

    public static void main(String[] args) {
        ThreadGroupDemo runnable = new ThreadGroupDemo();
        ThreadGroup tg1 = new ThreadGroup("Parent ThreadGroup");

        Thread t1 = new Thread(tg1, runnable, "one");
        t1.start();
        Thread t2 = new Thread(tg1, runnable, "two");
        t2.start();
        Thread t3 = new Thread(tg1, runnable, "three");
        t3.start();

        System.out.println("Thread Group Name: " + tg1.getName());
        tg1.list();
    }
}

ThreadGroupDemo.main(new String[]{})
```

  <h2> synchronization در جاوا </h2>



  <li>
همگام‌سازی یا <strong>Synchronization</strong> در جاوا قابلیتی است که به ما این امکان را می‌دهد که دسترسی چند نخ (Thread) را به یک منبع مشترک (<strong>Shared Resource</strong>) کنترل کنیم.  </li>
  </li>

  <li>
در واقع، وقتی چند نخ به‌طور هم‌ زمان به یک شیء یا متغیر مشترک دسترسی دارند، احتمال بروز تداخل (<strong>Interference</strong>) و ناهماهنگی داده‌ها (<strong>Inconsistency</strong>) وجود دارد.
  </li>

  <li>
با استفاده از همگام‌سازی، می‌توانیم این مشکل را برطرف کنیم تا در هر لحظه فقط یک نخ به منبع مورد نظر دسترسی داشته باشد.
  </li>

  <li>
<b>دلایل اصلی استفاده از همگام‌سازی در جاوا عبارت‌اند از:</b>
  </li>

  <p style="margin-bottom: 4px; text-indent: 35px;"> 
    <strong>1)</strong>    جلوگیری از تداخل نخ‌ها (Thread Interference):

  </p> 
  <p style="margin-bottom: 4px; text-indent: 55px;"> 
زمانی که چند نخ به‌طور هم‌زمان مقدار یک متغیر مشترک را تغییر می‌دهند، ممکن است مقدار نهایی اشتباه شود.
  </p> 
  <p style="margin-bottom: 4px; text-indent: 55px;"> 
همگام‌سازی تضمین می‌کند که تغییرات به‌صورت کنترل‌شده و پشت‌سر‌هم انجام شوند.
  </p> 

  <p style="margin-bottom: 4px; text-indent: 35px;"> 
    <strong>2)</strong>    حفظ سازگاری داده‌ها (Data Consistency):

  </p> 
  <p style="margin-bottom: 20px; text-indent: 55px;"> 
همگام‌سازی مانع می‌شود داده‌ها در اثر اجرای هم‌زمان نخ‌ها دچار بی‌نظمی یا ناهماهنگی شوند.
  </p> 
  
  <li>
  <b>همگام‌سازی در جاوا به دو دسته‌ی اصلی تقسیم می‌شود:</b>
  </li>
  <p style="margin-bottom: 4px; text-indent: 35px;"> 
    <strong>1)</strong>    انحصار متقابل (Mutual Exclusion)
  </p> 
  <p style="margin-bottom: 4px; text-indent: 55px;"> 
این نوع همگام‌سازی باعث می‌شود در یک زمان فقط یک نخ بتواند به بخش بحرانی (Critical Section) از کد دسترسی داشته باشد.
  </p> 
  <p style="margin-bottom: 4px; text-indent: 55px;"> 
روش‌های پیاده‌سازی آن:
  </p>

  <p style="margin-bottom: 4px; text-indent: 75px;"> 
<strong>1. Synchronized Method</strong> 
  </p> 
  <p style="margin-bottom: 4px; text-indent: 75px;"> 
<strong>2. Synchronized Block</strong> 
  </p> 
  <p style="margin-bottom: 4px; text-indent: 75px;"> 
<strong>3. Static Synchronization</strong> 
  </p> 

  
  <p style="margin-bottom: 4px; text-indent: 35px;"> 
    <strong>2)</strong>    همکاری بین نخ‌ها (Inter-thread Communication)
  </p> 
  <p style="margin-bottom: 20px; text-indent: 55px;"> 
این بخش از همگام‌سازی به ارتباط و هماهنگی بین نخ‌ها مربوط است، به‌طوری که نخ‌ها بتوانند با یکدیگر تبادل داده یا سیگنال انجام دهند.
  </p> 

</div>

  <h2> مفهوم Lock یا قفل در جاوا</h2>


  <li style="margin-bottom: 10px">
همگام‌سازی (Synchronization) در جاوا بر پایه‌ی مفهومی به نام <strong>قفل</strong> <strong>(Lock)</strong> یا <strong>مانیتور</strong> <strong>(Monitor) </strong> ساخته شده است.
  </li>

  <li style="margin-bottom: 10px">
در واقع، <strong>هر شیء</strong> در جاوا دارای یک قفل داخلی است که از آن برای کنترل دسترسی چند رشته (Thread) به منابع مشترک استفاده می‌شود.
  </li>

  <li style="margin-bottom: 10px">
زمانی که یک Thread بخواهد به داده‌های یک شیء به‌صورت هم‌زمان و ایمن دسترسی پیدا کند، باید ابتدا قفل آن شیء را به‌دست آورد (lock را بگیرد) و پس از پایان کار، قفل را آزاد کند تا نخ های دیگر بتوانند از آن استفاده کنند.
  </li>

<p style="text-indent: 35px; margin-bottom: 10px;">
به عبارت دیگر:
</p>

  <li style = "text-indent: 55px; margin-bottom: 20px">
تا زمانی که یک Thread قفل یک شیء را در اختیار دارد، هیچ Thread دیگری نمی‌تواند به بخش‌های هم‌زمان‌شده‌ی آن شیء دسترسی پیدا کند.
  </li>

  
  

</div>

<div class="card" dir="rtl">
  <h2>  مثال (بدون Synchronization) </h2>


  <p style="margin-bottom: 10px">
در مثال زیر، به دلیل نبود همگام‌سازی، چند Thread به‌طور هم‌زمان به منبع مشترک دسترسی دارند،
  </p>

  <p style="margin-bottom: 20px">
در نتیجه خروجی برنامه <b>ناهماهنگ و غیرقابل‌پیش‌بینی</b> خواهد بود:
  </p>


</div>


```java
class Table {
    void printTable(int n) { // method not synchronized
        for (int i = 1; i <= 5; i++) {
            System.out.println(n * i);
            try {
                Thread.sleep(400);
            } catch (Exception e) {
                System.out.println(e);
            }
        }
    }
}

class MyThread1 extends Thread {
    Table t;
    MyThread1(Table t) {
        this.t = t;
    }
    public void run() {
        t.printTable(5);
    }
}

class MyThread2 extends Thread {
    Table t;
    MyThread2(Table t) {
        this.t = t;
    }
    public void run() {
        t.printTable(100);
    }
}

public class TestSynchronization1 {
    public static void main(String args[]) {
        Table obj = new Table(); // shared resource (only one object)

        MyThread1 t1 = new MyThread1(obj);
        MyThread2 t2 = new MyThread2(obj);

        t1.start();
        t2.start();
    }
}
```

  <h2>  متد synchronized در جاوا </h2>

  <li style="margin-bottom: 10px">
اگر متدی را در جاوا با کلیدواژه‌ی <strong>synchronized</strong> تعریف کنیم، به آن متد همگام‌سازی‌شده گفته می‌شود.
  </li>

<p style="text-indent: 35px; margin-bottom: 10px;">
به عبارت دیگر:
</p>

  <li style = "text-indent: 55px; margin-bottom: 20px">
زمانی که یک متد synchronized باشد، <strong>فقط یک</strong> Thread در هر لحظه می‌تواند به آن متد از طریق همان شیء دسترسی داشته باشد.
  </li>

<li style="margin-bottom: 20px">
    متدهای همگام‌سازی‌شده برای <strong>قفل کردن</strong> شیء (Object Lock) هنگام دسترسی به منابع مشترک استفاده می‌شوند.
</li>

<li>
  وقتی یک Thread متد synchronized را فراخوانی می‌کند:
  </li>
  <p style="margin-bottom: 6px; text-indent: 35px;"> 
    <strong>1. </strong>    ابتدا قفل آن شیء را در اختیار می‌گیرد (Lock acquisition)
  </p> 
  <p style="margin-bottom: 6px; text-indent: 35px;"> 
    <strong>2. </strong>    تا زمانی که متد تمام نشده، هیچ Thread دیگری نمی‌تواند وارد همان متد (یا متد هم‌زمان‌شده‌ی دیگری از همان شیء) شود
  </p> 
  <p style="margin-bottom: 20px; text-indent: 35px;"> 
    <strong>3. </strong>    پس از اتمام متد، قفل آزاد می‌شود (Lock release)
  </p> 



</div>


```java
class Table {
    synchronized void printTable(int n) { // synchronized method
        for (int i = 1; i <= 5; i++) {
            System.out.println(n * i);
            try {
                Thread.sleep(400);
            } catch (Exception e) {
                System.out.println(e);
            }
        }
    }
}

class MyThread1 extends Thread {
    Table t;
    MyThread1(Table t) {
        this.t = t;
    }
    public void run() {
        t.printTable(5);
    }
}

class MyThread2 extends Thread {
    Table t;
    MyThread2(Table t) {
        this.t = t;
    }
    public void run() {
        t.printTable(100);
    }
}

public class TestSynchronization2 {
    public static void main(String args[]) {
        Table obj = new Table(); // shared resource (only one object)

        MyThread1 t1 = new MyThread1(obj);
        MyThread2 t2 = new MyThread2(obj);

        t1.start();
        t2.start();
    }
}
```

  <h2>  synchronized block در جاوا </h2>

  <li style="margin-bottom: 10px">
بلوک همگام‌سازی‌شده <strong>(Synchronized Block)</strong> در جاوا برای کنترل دسترسی چندین ترد به بخش خاصی از کد استفاده می‌شود.  </li>

  <li style = "margin-bottom: 30px">
برخلاف متد همگام‌سازی‌شده که کل متد را قفل می‌کند، در بلوک همگام‌سازی‌شده می‌توان فقط <strong>بخش مشخصی از کد</strong> را قفل کرد.
  </li>

<p style="margin-bottom: 10px">
    فرض کنید متد شما شامل ۵۰ خط کد است، اما فقط ۵ خط از آن نیاز به همگام‌سازی دارد.
</p>
<p style="margin-bottom: 30px">
    در این حالت، با استفاده از Synchronized Block می‌توانید فقط همان بخش خاص را قفل کنید تا سرعت اجرای برنامه کاهش پیدا نکند.
</p>

<li>
  <b>نکات مهم برای Synchronized Block:</b>
  </li>
  <p style="margin-bottom: 6px; text-indent: 35px;"> 
    <strong>1. </strong>    بلوک همگام‌سازی‌شده برای قفل کردن یک شیء خاص در زمان دسترسی چند ترد به منبع مشترک استفاده می‌شود.
  </p> 
  <p style="margin-bottom: 6px; text-indent: 35px;"> 
    <strong>2. </strong>    محدوده‌ی (Scope) بلوک همگام‌سازی‌شده از متد همگام‌سازی‌شده کوچک‌تر است.
  </p> 
  <p style="margin-bottom: 30px; text-indent: 35px;"> 
    <strong>3. </strong>    استفاده از آن باعث می‌شود فقط بخش بحرانی (Critical Section) از کد قفل شود و بقیه قسمت‌ها آزاد بمانند.
  </p> 


<p dir='ltr' style="margin-bottom: 0.5px;">
  Syntax to use synchronized block : 
  </p>
  <pre style="background-color: #272822; color: #f8f8f2; padding: 10px; border-radius: 8px; direction: ltr; text-align: left;">
<span style="color: #22e1faff;">synchronized</span> (object reference expression) {
    <span style="color: #42ca20ff;">// code block</span>
}
</pre>



</div>


```java
class Table {  
    void printTable(int n) {  
        synchronized(this) {  // synchronized block
            for (int i = 1; i <= 5; i++) {  
                System.out.println(n * i);  
                try {  
                    Thread.sleep(400);  
                } catch (Exception e) {  
                    System.out.println(e);  
                }  
            }  
        }
    } // end of the method 
}  

class MyThread1 extends Thread {  
    Table t;  
    MyThread1(Table t) {  
        this.t = t;  
    }  
    public void run() {  
        t.printTable(5);  
    }  
}  

class MyThread2 extends Thread {  
    Table t;  
    MyThread2(Table t) {  
        this.t = t;  
    }  
    public void run() {  
        t.printTable(100);  
    }  
}  

public class TestSynchronizedBlock {  
    public static void main(String args[]) {  
        Table obj = new Table(); // shared resource (only one object)

        MyThread1 t1 = new MyThread1(obj);  
        MyThread2 t2 = new MyThread2(obj);  

        t1.start();  
        t2.start();  
    }  
}
```

  <h2> پایان دادن ترد ها در جاوا </h2>


  <li style="margin-bottom: 10px">
در جاوا، پایان دادن به نخ‌ها (Killing Threads) یکی از بخش‌های مهم در کنترل اجرای چندنخی (Multithreading) است.
  </li>

  <li style="margin-bottom: 10px">
به‌ صورت پیش‌ فرض، وقتی متد ()run در یک نخ تمام می‌شود، آن نخ به‌صورت خودکار از بین می‌رود.
  </li> 

  <li style="margin-bottom: 10px;">
  اما گاهی نیاز داریم نخ را قبل از اتمام اجرای کاملش متوقف یا حذف کنیم.
  </li>

<li>
  در نسخه‌های اولیه‌ی جاوا، برای کنترل اجرای نخ‌ها از متدهای زیر استفاده می‌شد:
  </li>
  <p style="margin-bottom: 6px; text-indent: 35px;"> 
    <strong>1. </strong>    ()suspend
  </p> 
  <p style="margin-bottom: 6px; text-indent: 35px;"> 
    <strong>2. </strong>    ()resume
  </p> 
  <p style="margin-bottom: 10px; text-indent: 35px;"> 
    <strong>3. </strong>    ()stop
  </p> 

  <p>
اما این متدها از نسخه‌ی Java 2 به بعد منسوخ (deprecated) شدند، چون می‌توانستند منجر به خرابی سیستم یا بی‌ثباتی داده‌ها شوند.
<br>
به‌عنوان مثال، اگر نخی در حال به‌روزرسانی داده‌ای مشترک بود و ناگهان متوقف می‌شد، داده در وضعیت ناپایدار باقی می‌ماند.
  </p>

<li>
  <b>روش‌ های مدرن برای توقف نخ‌ها :</b>
  </li>
  <p style="margin-bottom: 6px; text-indent: 35px;"> 
    <strong>1. استفاده از پرچم  (Boolean Flag)</strong>   
  </p> 
  <p style="margin-bottom: 6px; text-indent: 55px;"> 
    در این روش، متغیری از نوع boolean (مثلاً با نام exit) تعریف می‌شود که نشان می‌دهد نخ باید ادامه دهد یا متوقف شود.
  </p> 
  <p style="margin-bottom: 6px; text-indent: 35px;"> 
    <strong>2. استفاده از ()Thread.interrupt</strong>    
  </p>
  <p style="margin-bottom: 6px; text-indent: 55px;"> 
روش دیگر برای توقف یک نخ، استفاده از متد ()interrupt است.
  </p> 
  <p style="margin-bottom: 20px; text-indent: 55px;"> 
این متد باعث نمی‌شود نخ بلافاصله متوقف شود، بلکه سیگنالی برای قطع اجرا ارسال می‌کند.
  </p> 

</div>


```java
// Example of killing threads in Java
class MyThread implements Runnable {

    // to stop the thread
    private boolean exit;
    private String name;
    Thread t;

    MyThread(String threadname) {
        name = threadname;
        t = new Thread(this, name);
        System.out.println("New thread: " + t);
        exit = false;
        t.start(); // Starting the thread
    }

    // execution of thread starts from run() method
    public void run() {
        int i = 0;
        while (!exit) {
            System.out.println(name + ": " + i);
            i++;
            try {
                Thread.sleep(100);
            } catch (InterruptedException e) {
                System.out.println("Caught: " + e);
            }
        }
        System.out.println(name + " Stopped.");
    }

    // for stopping the thread
    public void stop() {
        exit = true;
    }
}

// Main class
public class Main {
    public static void main(String args[]) {
        // creating two objects t1 & t2 of MyThread
        MyThread t1 = new MyThread("First thread");
        MyThread t2 = new MyThread("Second thread");

        try {
            Thread.sleep(500);
            t1.stop(); // stopping thread t1
            t2.stop(); // stopping thread t2
            Thread.sleep(500);
        } catch (InterruptedException e) {
            System.out.println("Caught: " + e);
        }

        System.out.println("Exiting the main Thread");
    }
}

Main.main(new String[]{});
```
