  <h2>مقدمه‌ای بر Set و Map</h2>
  <p>
    در این بخش، مسیر خود را در چارچوب <strong>Collections جاوا</strong> ادامه می‌دهیم و روی دو واسط بسیار مهم تمرکز می‌کنیم: <strong>Set</strong> و <strong>Map</strong>. هدف ما درک تفاوت‌های ساختاری و کاربردی این کالکشن‌هاست.
  </p>
  <p>
    ابتدا با واسط <code>Set</code> آشنا می‌شویم که مفهوم مجموعه با عناصر <strong>منحصر به فرد و بدون تکرار</strong> را پیاده‌سازی می‌کند. سپس به سراغ واسط <code>Map</code> می‌رویم که برای ذخیره‌سازی داده‌ها به صورت جفت‌های <strong>کلید-مقدار (Key-Value)</strong> به کار می‌رود. درک این ساختارها به ما کمک می‌کند تا برای هر سناریو، بهترین و کارآمدترین ابزار را انتخاب کنیم.
  </p>

  <h2>مروری بر واسط Set</h2>
  <ul>
    <li>یک <code>Set</code> در جاوا معادل مفهوم <strong>مجموعه در ریاضیات</strong> است که از ورود عناصر تکراری جلوگیری می‌کند.</li>
    <li>مهم‌ترین ویژگی <code>Set</code>، اهمیت دادن به <strong>منحصر به فرد بودن (uniqueness)</strong> عناصر است و اجازه تکرار نمی‌دهد.</li>
    <li><code>Set</code> یک واسط (Interface) است و نمی‌توان آن را مستقیماً نمونه‌سازی کرد (یعنی نمی‌توان گفت <code>new Set()</code>).</li>
    <li>برای استفاده از آن باید یکی از کلاس‌های پیاده‌ساز مانند <code>HashSet</code>، <code>LinkedHashSet</code> یا <code>TreeSet</code> را به کار برد.</li>
  </ul>

<div class="card" dir="rtl">
  <h2>تصادم‌ها (Collisions) در جدول هش</h2>
  <p>
    یک چالش مهم در Hashing زمانی رخ می‌دهد که تابع هش برای دو یا چند شیء متفاوت، یک کد هش <strong>یکسان</strong> تولید کند. به این پدیده <strong>تصادم (Collision)</strong> گفته می‌شود. برای مثال، دو رشته متفاوت `"Aa"` و `"BB"` می‌توانند کد هش یکسانی داشته باشند.
  </p>
  <p>
    رایج‌ترین راه‌حل برای مدیریت تصادم، استفاده از روش <strong>جدول هش زنجیره‌ای (Separate Chaining)</strong> است. در این روش، هر خانه از جدول هش به یک لیست پیوندی (Linked List) اشاره می‌کند. اگر چند عنصر دارای کد هش یکسان باشند، همه آن‌ها در لیست پیوندی مربوط به آن خانه ذخیره می‌شوند.
  </p>
  <ul>
    <li>این روش تضمین می‌کند که هیچ داده‌ای از بین نرود.</li>
    <li>البته، در صورت وقوع تصادم‌های زیاد، زمان جستجو کمی بیشتر می‌شود، زیرا ممکن است نیاز به پیمایش یک لیست کوتاه باشد.</li>
    <li>به همین دلیل است که ساختار داده <code>HashSet</code> ترتیب ورود عناصر را حفظ نمی‌کند؛ زیرا جایگاه هر عنصر توسط کد هش آن تعیین می‌شود.</li>
  </ul>

  </div>
  <div class="slide-note">
    <strong>نکته:</strong> یک تابع هش ایده‌آل که هیچ‌گاه تصادم ایجاد نمی‌کند، "تابع هش بی‌عیب" (Perfect Hash Function) نامیده می‌شود، اما در عمل رسیدن به آن دشوار است.

<div class="card" dir="rtl">
  <h2>پیاده‌سازی Set: کلاس HashSet</h2>
  <p>
    <code>HashSet</code> رایج‌ترین پیاده‌سازی واسط <code>Set</code> است. این کلاس از یک جدول هش (Hash Table) برای ذخیره‌سازی عناصر خود استفاده می‌کند. مهم‌ترین ویژگی‌های آن سرعت بالا و عدم تضمین ترتیب عناصر است.
  </p>
  <p>
    به لطف استفاده از جدول هش، عملکرد متدهای اصلی مانند <code>add</code>، <code>remove</code> و <code>contains</code> به طور میانگین دارای پیچیدگی زمانی ثابت <strong>O(1)</strong> است که آن را برای کاربردهایی که سرعت در آن‌ها اهمیت بالایی دارد، ایده‌آل می‌سازد.
  </p>
  <ul>
    <li>عناصر در <code>HashSet</code> بر اساس کد هش (hash code) خود ذخیره می‌شوند.</li>
    <li>این کلاس اجازه ورود عناصر تکراری را نمی‌دهد.</li>
    <li>ترتیب پیمایش عناصر در <code>HashSet</code> تضمین‌شده نیست و ممکن است تصادفی به نظر برسد.</li>
    <li>برای اینکه جاوا بفهمد یک عنصر در مجموعه وجود دارد، ابتدا <code>hashCode()</code> آن را بررسی می‌کند و سپس از متد <code>equals()</code> برای مقایسه نهایی استفاده می‌کند.</li>
  </ul>
  <div class="images">
    <img src="img/hashse.png" alt="HashSet Internal Storage Diagram" >
  </div>
  <div class="slide-note">
    <strong>چه زمانی استفاده کنیم؟</strong> هرگاه به یک مجموعه بدون عناصر تکراری نیاز داشتید و ترتیب عناصر برایتان اهمیتی نداشت، <code>HashSet</code> بهترین انتخاب است.

<div class="card" dir="rtl">
  <h2>پیاده‌سازی Set: کلاس LinkedHashSet</h2>
  <p>
    کلاس <code>LinkedHashSet</code> یک پیاده‌سازی خاص از <code>Set</code> است که علاوه بر تمام ویژگی‌های <code>HashSet</code>، <strong>ترتیب ورود عناصر</strong> را نیز حفظ می‌کند. این کلاس ساختاری مابین <code>HashSet</code> و <code>TreeSet</code> دارد.
  </p>
  <p>
    برای رسیدن به این هدف، <code>LinkedHashSet</code> علاوه بر جدول هش، یک <strong>لیست پیوندی دوطرفه (Doubly Linked List)</strong> را نیز در پس‌زمینه نگهداری می‌کند. جدول هش برای اطمینان از سرعت O(1) در عملیات اصلی به کار می‌رود، در حالی که لیست پیوندی وظیفه حفظ ترتیب افزودن عناصر را بر عهده دارد.
  </p>
  <ul>
    <li>مانند <code>HashSet</code>، سرعت عملیات <code>add</code>، <code>remove</code> و <code>contains</code> برابر O(1) است.</li>
    <li>برخلاف <code>HashSet</code>، هنگام پیمایش این مجموعه، عناصر به همان ترتیبی که وارد شده‌اند، بازیابی می‌شوند.</li>
    <li>این کلاس نیز مانند <code>HashSet</code> تنها عناصر منحصر به فرد را می‌پذیرد.</li>
  </ul>
  <div class="images">
    <img src="img/linkedhashset_structure.png" alt="LinkedHashSet Internal Structure Diagram" style="width: 100%; max-width: 900px; height: auto;">
  </div>
  <div class="slide-note">
    <strong>چه زمانی استفاده کنیم؟</strong> زمانی که به ویژگی‌های یک <code>Set</code> (عدم تکرار و سرعت بالا) نیاز دارید، اما ترتیب افزودن عناصر نیز برای شما مهم است، <code>LinkedHashSet</code> بهترین گزینه است.

  <h3>آزمون کارایی</h3>
  <p>
    برای درک بهتر تفاوت عملکرد، می‌توانیم یک آزمون ساده اجرا کنیم. در کد زیر، به هر سه نوع <code>Set</code> تعداد زیادی عنصر اضافه کرده و زمان مورد نیاز را اندازه‌گیری می‌کنیم. همانطور که مشاهده می‌شود، <code>TreeSet</code> به دلیل سربار مرتب‌سازی، کندتر از دو گزینه دیگر عمل می‌کند.
  </p>

import java.util.*;

public class PerformanceTest {
    public static void main(String[] args) {
        Random r = new Random();
        HashSet<Dog> hashSet = new HashSet<>();
        TreeSet<Dog> treeSet = new TreeSet<>();
        LinkedHashSet<Dog> linkedSet = new LinkedHashSet<>();

        // --- Test HashSet ---
        long startTime = System.nanoTime();
        for (int i = 0; i < 1000; i++) {
            hashSet.add(new Dog(r.nextInt(1000)));
        }
        long duration = System.nanoTime() - startTime;
        System.out.println("HashSet: " + duration);

        // --- Test TreeSet ---
        startTime = System.nanoTime();
        for (int i = 0; i < 1000; i++) {
            treeSet.add(new Dog(r.nextInt(1000)));
        }
        duration = System.nanoTime() - startTime;
        System.out.println("TreeSet: " + duration);

        // --- Test LinkedHashSet ---
        startTime = System.nanoTime();
        for (int i = 0; i < 1000; i++) {
            linkedSet.add(new Dog(r.nextInt(1000)));
        }
        duration = System.nanoTime() - startTime;
        System.out.println("LinkedHashSet: " + duration);
    }
    // (Dog class must implement Comparable for TreeSet to work)
    static class Dog implements Comparable<Dog> {
        int size;
        Dog(int s) { size = s; }
        @Override
        public int compareTo(Dog o) { return this.size - o.size; }
    }
}

  <h2>تفاوت میان Set و List</h2>
  <p>
    اگرچه هر دو واسط <code>Set</code> و <code>List</code> از <code>Collection</code> ارث‌بری می‌کنند، اما برای اهداف متفاوتی طراحی شده‌اند و تفاوت‌های بنیادینی با یکدیگر دارند. درک این تفاوت‌ها برای انتخاب ساختمان داده مناسب، ضروری است.
  </p>
  
  <ul>
    <li>
      <strong>عناصر تکراری:</strong> مهم‌ترین تفاوت این است که <code>List</code> اجازه ورود عناصر تکراری را می‌دهد، در حالی که <code>Set</code> تنها عناصر <strong>منحصر به فرد</strong> را نگهداری می‌کند.
    </li>
    <li>
      <strong>ترتیب عناصر:</strong> <code>List</code> یک کالکشن مرتب است و ترتیب افزودن عناصر را حفظ می‌کند. اما <code>Set</code> به طور کلی یک کالکشن بدون ترتیب است (به جز <code>LinkedHashSet</code> که ترتیب ورود و <code>TreeSet</code> که ترتیب طبیعی را حفظ می‌کند).
    </li>
    <li>
      <strong>مقادیر Null:</strong> یک <code>List</code> می‌تواند هر تعداد مقدار <code>null</code> را در خود جای دهد، اما <code>Set</code> حداکثر یک مقدار <code>null</code> را می‌پذیرد.
    </li>
  </ul>

  <h3>چه زمانی از کدام استفاده کنیم؟</h3>
  <ul>
    <li>اگر می‌خواهید ترتیب عناصر حفظ شود و ممکن است عناصر تکراری داشته باشید، از <strong>List</strong> استفاده کنید.</li>
    <li>اگر می‌خواهید مطمئن شوید که هیچ عنصر تکراری در مجموعه وجود ندارد و ترتیب برایتان مهم نیست، از <strong>Set</strong> استفاده کنید.</li>
    <li>اگر هم به منحصر به فرد بودن و هم به حفظ ترتیب ورود نیاز دارید، از <strong>LinkedHashSet</strong> استفاده کنید.</li>
  </ul>

  <p>کد زیر تفاوت در عمل را نشان می‌دهد:</p>
  <pre dir="ltr" style="text-align: left; background-color: black; color: white; border: 1px solid #ccc; border-radius: 4px; padding: 10px; direction: ltr;"><code class="language-java">
import java.util.ArrayList;
import java.util.HashSet;
import java.util.List;
import java.util.Set;

    public class ListVsSetExample {
        public static void main(String[] args) {
            // List Example
            List<String> list = new ArrayList<>();
            list.add("Apple");
            list.add("Banana");
            list.add("Apple"); // Duplicate is allowed
            System.out.println("List: " + list);
            // Output: List: [Apple, Banana, Apple] (Order is preserved, duplicates allowed)

            // Set Example
            Set<String> set = new HashSet<>();
            set.add("Apple");
            set.add("Banana");
            set.add("Apple"); // Duplicate is ignored
            System.out.println("Set: " + set);
            // Output: Set: [Apple, Banana] (Order is not guaranteed, duplicates are not allowed)
        }
    }

<div class="card" dir="rtl">
  <h2>آشنایی با واسط Map</h2>
  <p>
    تا اینجا کالکشن‌هایی را دیدیم که مجموعه‌ای از عناصر را نگهداری می‌کردند. اما گاهی نیاز داریم تا یک نگاشت یا ارتباط یک-به-یک بین دو مجموعه از داده‌ها برقرار کنیم. برای مثال، لیستی از نام افراد (کلید) و شماره تلفن آن‌ها (مقدار). واسط <code>Map</code> دقیقاً برای همین منظور طراحی شده است.
  </p>
  <p>
    یک <code>Map</code> یک ساختمان داده است که داده‌ها را به صورت جفت‌های <strong>کلید-مقدار (Key-Value)</strong> ذخیره می‌کند. ویژگی اصلی آن این است که هر کلید باید <strong>منحصر به فرد</strong> باشد و از طریق آن می‌توان به سرعت به مقدار متناظرش دسترسی پیدا کرد.
  </p>
  <ul>
    <li>
      <strong>کلید (Key):</strong> یک شناسه منحصر به فرد برای هر مقدار است.
    </li>
    <li>
      <strong>مقدار (Value):</strong> داده‌ای است که به یک کلید خاص مرتبط شده است. مقادیر می‌توانند تکراری باشند.
    </li>
    <li>
      عملیات اصلی در <code>Map</code> شامل <code>put(key, value)</code> برای افزودن یا به‌روزرسانی یک جفت، و <code>get(key)</code> برای بازیابی مقدار با استفاده از کلید آن است.
    </li>
  </ul>
  
  <p>به طور خلاصه، تفاوت اصلی در نوع پرسشی است که از هر ساختار می‌پرسیم:</p>
  <ul>
    <li><strong>List:</strong> "آیتم در اندیس ۵ چیست؟"</li>
    <li><strong>Set:</strong> "آیا آیتم 'Apple' موجود است؟"</li>
    <li><strong>Map:</strong> "مقدار مرتبط با کلید 'Apple' چیست؟"</li>

  <h2>پیاده‌سازی Map: کلاس HashMap</h2>
  <p>
    <code>HashMap</code> پرکاربردترین پیاده‌سازی واسط <code>Map</code> در جاوا است. این کلاس از یک جدول هش (Hash Table) برای ذخیره جفت‌های کلید-مقدار استفاده می‌کند که باعث می‌شود عملکرد آن برای عملیات اصلی بسیار سریع باشد.
  </p>
  <p>
    همانند <code>HashSet</code>، عملکرد متدهای <code>put</code> (افزودن) و <code>get</code> (بازیابی) در <code>HashMap</code> به طور میانگین دارای پیچیدگی زمانی <strong>O(1)</strong> است. همچنین، <code>HashMap</code> هیچ تضمینی برای حفظ ترتیب کلیدها نمی‌دهد و ترتیب پیمایش آن ممکن است تصادفی باشد.
  </p>
  
  <h3>اهمیت متدهای equals() و hashCode()</h3>
  <p>
    عملکرد صحیح <code>HashMap</code> به شدت به پیاده‌سازی متدهای <code>equals()</code> و <code>hashCode()</code> در کلاسِ کلیدها بستگی دارد:
  </p>
  <ul>
    <li>
      وقتی یک جفت کلید-مقدار را با <code>put()</code> اضافه می‌کنید، <code>HashMap</code> از <code>key.hashCode()</code> برای پیدا کردن خانه (bucket) مناسب در جدول هش استفاده می‌کند.
    </li>
    <li>
      هنگام بازیابی با <code>get(key)</code>، ابتدا با <code>hashCode()</code> خانه مربوطه را پیدا کرده و سپس با متد <code>equals()</code> کلید دقیق را در میان تمام کلیدهای آن خانه جستجو می‌کند.
    </li>
    <li>
      <strong>قانون طلایی:</strong> اگر از اشیاء یک کلاس سفارشی به عنوان کلید در <code>HashMap</code> استفاده می‌کنید، <strong>باید</strong> حتماً هر دو متد <code>equals()</code> و <code>hashCode()</code> را به درستی بازنویسی (override) کنید.
    </li>
  </ul>
  
  <pre dir="ltr" style="text-align: left; background-color: black; color: white; border: 1px solid #ccc; border-radius: 4px; padding: 10px; direction: ltr;"><code class="language-java">
import java.util.HashMap;
import java.util.Map;

public class MyHashMapExample {
    public static void main(String[] args) {
        // Create a HashMap
        Map<String, Integer> grades = new HashMap<>();

        // Add key-value pairs
        grades.put("Martin", 85);
        grades.put("Nelson", 45);
        grades.put("Milhouse", 70);

        // Get a value by its key
        System.out.println("Nelson's grade: " + grades.get("Nelson"));

        // Update a value
        grades.put("Nelson", 55);
        System.out.println("Nelson's new grade: " + grades.get("Nelson"));

        // Remove a key-value pair
        grades.remove("Martin");
        
        // Iterate over the map
        for (Map.Entry<String, Integer> entry : grades.entrySet()) {
            System.out.println(entry.getKey() + ": " + entry.getValue());
        }
    }
}
  </code></pre>

  <h2>سایر پیاده‌سازی‌های Map: LinkedHashMap و TreeMap</h2>
  <p>
    همانند <code>Set</code>، واسط <code>Map</code> نیز پیاده‌سازی‌هایی دارد که ترتیب کلیدها را تضمین می‌کنند. اگر ترتیب برای شما مهم است، می‌توانید به جای <code>HashMap</code> از <code>LinkedHashMap</code> یا <code>TreeMap</code> استفاده کنید.
  </p>
  
  <h3>کلاس LinkedHashMap</h3>
  <p>
    این کلاس تقریباً مشابه <code>HashMap</code> عمل می‌کند اما یک ویژگی مهم دارد: ترتیب افزودن کلیدها را حفظ می‌کند. <code>LinkedHashMap</code> این کار را با استفاده از یک لیست پیوندی دوطرفه که در پس‌زمینه نگهداری می‌شود، انجام می‌دهد. عملکرد آن همچنان <strong>O(1)</strong> است.
  </p>

  <h3>کلاس TreeMap</h3>
  <p>
    این کلاس جفت‌های کلید-مقدار را بر اساس <strong>ترتیب طبیعی کلیدها (sorted order)</strong> نگهداری می‌کند. <code>TreeMap</code> با استفاده از یک درخت سرخ-سیاه پیاده‌سازی شده و عملکرد آن <strong>O(log n)</strong> است. کلیدهایی که در <code>TreeMap</code> استفاده می‌شوند، حتماً باید واسط <code>Comparable</code> را پیاده‌سازی کرده باشند.
  </p>

  <p>کد زیر تفاوت ترتیب در این دو کلاس را نشان می‌دهد:</p>
  <pre dir="ltr" style="text-align: left; background-color: black; color: white; border: 1px solid #ccc; border-radius: 4px; padding: 10px; direction: ltr;"><code class="language-java">
import java.util.LinkedHashMap;
import java.util.Map;
import java.util.TreeMap;

    public class OrderedMapsExample {
        public static void main(String[] args) {
            // LinkedHashMap maintains insertion order
            Map<Integer, String> linkedHashMap = new LinkedHashMap<>();
            linkedHashMap.put(22, "Abey");
            linkedHashMap.put(33, "Dawn");
            linkedHashMap.put(1, "Sherry");
            System.out.println("LinkedHashMap (Insertion Order): " + linkedHashMap);
            // Output: {22=Abey, 33=Dawn, 1=Sherry}

            // TreeMap maintains natural sorting order of keys
            Map<Integer, String> treeMap = new TreeMap<>();
            treeMap.put(22, "Abey");
            treeMap.put(33, "Dawn");
            treeMap.put(1, "Sherry");
            System.out.println("TreeMap (Sorted Order): " + treeMap);
            // Output: {1=Sherry, 22=Abey, 33=Dawn}
        }
    }
  </code></pre>
  
  <div class="slide-note">
    <strong>جمع‌بندی نهایی:</strong> برای انتخاب <code>Map</code> مناسب، به نیاز خود توجه کنید:
    <ul>
      <li><strong>HashMap:</strong> برای حداکثر سرعت و زمانی که ترتیب مهم نیست.</li>
      <li><strong>LinkedHashMap:</strong> برای زمانی که به سرعت HashMap و حفظ ترتیب ورود نیاز دارید.</li>
      <li><strong>TreeMap:</strong> برای زمانی که نیاز دارید کلیدها همیشه مرتب باشند.</li>
    </ul>
  </div>

  <h3 style="color: #dc3545;">سطح سخت (Hard)</h3>

  <p><strong>سوال ۱:</strong> یک کلاس <code>Employee</code> با فیلدهای <code>id</code> و <code>name</code> بسازید. متدهای <code>equals()</code> و <code>hashCode()</code> را به درستی بازنویسی کنید تا دو کارمند با <code>id</code> یکسان، برابر در نظر گرفته شوند. سپس از این کلاس به عنوان کلید در یک <code>HashMap</code> استفاده کنید و نشان دهید که اگر یک کارمند با <code>id</code> تکراری اضافه کنید، مقدار قبلی بازنویسی می‌شود.</p>
  <p><em>پاسخ:</em></p>
  <pre dir="ltr" style="text-align: left; background-color: black; border: 1px solid #555; border-radius: 4px; padding: 10px; direction: ltr;"><code class="language-java" style="color: white; font-weight: 900;">
import java.util.HashMap;
import java.util.Map;
import java.util.Objects;

    class Employee {
        int id;
        String name;

        public Employee(int id, String name) { this.id = id; this.name = name; }
    
        @Override
        public boolean equals(Object o) {
            if (this == o) return true;
            if (o == null || getClass() != o.getClass()) return false;
            Employee employee = (Employee) o;
            return id == employee.id;
        }

        @Override
        public int hashCode() {
            return Objects.hash(id);
        }
    
        @Override
        public String toString() { return "ID: " + id + ", Name: " + name; }
    }

    public class CustomKeyHashMap {
        public static void main(String[] args) {
            Map<Employee, String> departments = new HashMap<>();
            departments.put(new Employee(101, "Alice"), "Engineering");
            departments.put(new Employee(102, "Bob"), "Marketing");
        
            // This will overwrite the value for employee with ID 101
            departments.put(new Employee(101, "Alice Smith"), "Technology");
        
            System.out.println("Map size: " + departments.size()); // Should be 2
            System.out.println(departments);
        }
    }
  </code></pre>

  <p><strong>سوال ۲:</strong> برنامه‌ای بنویسید که کارآمدترین مسیر پرواز بین دو شهر را بر اساس کمترین تعداد توقف پیدا کند. از یک <code>Map<String, List<String>></code> برای نمایش گراف پروازها و از الگوریتم جستجوی سطح اول (BFS) استفاده کنید.</p>
  <p><em>پاسخ:</em></p>
  <pre dir="ltr" style="text-align: left; background-color: black; border: 1px solid #555; border-radius: 4px; padding: 10px; direction: ltr;"><code class="language-java" style="color: white; font-weight: 900;">
import java.util.*;

    public class FlightPathFinder {
        public static void main(String[] args) {
            Map<String, List<String>> flights = new HashMap<>();
            flights.put("JFK", Arrays.asList("LAX", "SFO"));
            flights.put("LAX", Arrays.asList("SFO", "HNL"));
            flights.put("SFO", Arrays.asList("HNL"));

            System.out.println("Shortest path from JFK to HNL: " + findShortestPath("JFK", "HNL", flights));
        }

        public static List<String> findShortestPath(String start, String end, Map<String, List<String>> graph) {
            Queue<List<String>> queue = new LinkedList<>();
            Set<String> visited = new HashSet<>();
            queue.add(Arrays.asList(start));
            visited.add(start);

            while (!queue.isEmpty()) {
                List<String> path = queue.poll();
                String current = path.get(path.size() - 1);
                if (current.equals(end)) return path;

                if (graph.containsKey(current)) {
                    for (String neighbor : graph.get(current)) {
                        if (!visited.contains(neighbor)) {
                            visited.add(neighbor);
                            List<String> newPath = new ArrayList<>(path);
                            newPath.add(neighbor);
                            queue.add(newPath);
                        }
                    }
                }
            }
            return null; // No path found
        }
    }
  </code></pre>

  <p><strong>سوال ۳:</strong> یک کلاس برای پیاده‌سازی <strong>نقشه دوطرفه (Bi-directional Map)</strong> بنویسید که بتوان با کلید، مقدار را پیدا کرد و با مقدار، کلید را. از دو <code>HashMap</code> داخلی برای این کار استفاده کنید.</p>

import java.util.HashMap;
import java.util.Map;

    class BiMap<K, V> {
        private Map<K, V> keyToValue = new HashMap<>();
        private Map<V, K> valueToKey = new HashMap<>();

        public void put(K key, V value) {
            keyToValue.put(key, value);
            valueToKey.put(value, key);
        }
        public V getByKey(K key) { return keyToValue.get(key); }
        public K getByValue(V value) { return valueToKey.get(value); }
    }

    public class BiMapTest {

            System.out.println("Role of user1: " + userRoles.getByKey("user1"));
            System.out.println("User with Admin role: " + userRoles.getByValue("Admin"));
        }
    }

  <p><strong>سوال ۴:</strong> یک <strong>کش LRU (Least Recently Used)</strong> ساده با استفاده از <code>LinkedHashMap</code> پیاده‌سازی کنید. این کش باید یک ظرفیت مشخص داشته باشد و وقتی پر شد، قدیمی‌ترین عنصر استفاده‌نشده را حذف کند.</p>
  <p><em>پاسخ:</em></p>
  <pre dir="ltr" style="text-align: left; background-color: black; border: 1px solid #555; border-radius: 4px; padding: 10px; direction: ltr;"><code class="language-java" style="color: white; font-weight: 900;">
import java.util.LinkedHashMap;
import java.util.Map;

    class LRUCache<K, V> extends LinkedHashMap<K, V> {
        private final int capacity;

        public LRUCache(int capacity) {
            super(capacity, 0.75f, true); // true for access-order
            this.capacity = capacity;
        }

        @Override
        protected boolean removeEldestEntry(Map.Entry<K, V> eldest) {
            return size() > capacity;
        }
    }

    public class LRUCacheTest {
        public static void main(String[] args) {
            LRUCache<Integer, String> cache = new LRUCache<>(2);
            cache.put(1, "A");
            cache.put(2, "B");
            System.out.println(cache); // {1=A, 2=B}
            cache.get(1); // Access 1, making it most recently used
            System.out.println(cache); // {2=B, 1=A}
            cache.put(3, "C"); // Cache is full, 2 (least recent) is removed
            System.out.println(cache); // {1=A, 3=C}
        }
    }
