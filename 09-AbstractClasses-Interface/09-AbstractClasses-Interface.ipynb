```python
abstract class Bike{
    abstract void run();
}

class Honda4 extends Bike{
    void run(){System.out.println("running safely...");}
    
    public static void main(String args[]){
        Bike obj = new Honda4();
        obj.run();
    }
}
```

```python
abstract class Shape{
    abstract void draw();
}
//In real scenario, implementation is provided by others i.e. unknown by end user
class Rectangle extends Shape{
    void draw(){System.out.println("drawing rectangle");}
}

class Circle1 extends Shape{
    void draw(){System.out.println("drawing circle");}
}

//In real scenario, method is called by programmer or user
class TestAbstraction1{
    public static void main(String args[]){
        Shape s=new Circle1();
        s.draw();
    }
}
```

```python
abstract class Bank{
    abstract int getRateOfInterest();
}

class SBI extends Bank{
    int getRateOfInterest(){return 7;}
}

class PNB extends Bank{
    int getRateOfInterest(){return 6;}
}

class TestBank{
    public static void main(String args[]){
        Bank b=new SBI();//If object is PNB, method of PNB will be invoked
        int interest=b.getRateOfInterest();
        System.out.println("Rate of Interest is: "+interest+" %");
    }
}
```

  <h3>متد abstract در کلاس abstract</h3>
  <p>از یک <strong>کلاس abstract</strong> با استفاده از کلیدواژه <strong>new</strong> نمی‌توان شی جدیدی ایجاد کرد اما همچنان می‌توانیم از <strong>سازنده</strong> این کلاس‌های انتزاعی در کلاس‌های فرزندشان استفاده کنیم.</p>

  <p>این کد یک ساختار شیءگرا را نشان می‌دهد که از مفاهیم کلاس انتزاعی و وراثت استفاده می‌کند.</p>


  <p>• یک کلاس انتزاعی به نام Animal تعریف شده است.<br>
     • این کلاس دارای یک متغیر نمونه به نام name و یک سازنده است.<br>
     • کلاس شامل یک متد انتزاعی makeSound بدون پیاده‌سازی می‌باشد.</p>

  <p>• یک کلاس معمولی به نام Dog از کلاس Animal ارث‌بری می‌کند.<br>
     • کلاس Dog سازنده‌ای دارد که سازنده کلاس والد را فراخوانی می‌کند.<br>
     • این کلاس متد makeSound را پیاده‌سازی کرده و بدنه آن را ارائه می‌دهد.</p>

  <p>• متد makeSound نام حیوان و صدای مربوطه را در خروجی چاپ می‌کند.<br>
     • اگرچه نمی‌توان از کلاس Animal مستقیماً شیء ایجاد کرد.<br>
     • اما سازنده آن در کلاس‌های فرزند قابل استفاده است.<br>
     • این ویژگی امکان قرار دادن کدهای مشترک در سازنده کلاس والد را فراهم می‌کند.</p>

  <hr class="divider" />
</div>

```python
abstract class Animal {
    String name;
    
    public Animal(String name) {
        this.name = name;
    }
    
    abstract void makeSound();
}

class Dog extends Animal {
    public Dog(String name) {
        super(name);
    }
    
    void makeSound() {
        System.out.println(name + " says: Woof!");
    }
}
```

  <h3>نکته مهم درباره کلاس‌های انتزاعی</h3>
  <p><strong>متد abstract</strong> نمی‌تواند درون یک <strong>کلاس غیر-abstract</strong> قرار گیرد.  
  این به این معنی است که اگر کلاسی قرار است نمونه‌سازی شود و به‌صورت کامل رفتار مشخصی داشته باشد، نباید شامل متدی باشد که هنوز <strong>بدون پیاده‌سازی</strong> است.  
  قرار دادن متد abstract در کلاس عادی باعث می‌شود آن کلاس ناقص باشد و نتوان از آن به‌طور مستقیم شیء ساخت.</p>

  <p>اگر یک <strong>زیرکلاس</strong> از یک <strong>ابرکلاس abstract</strong>، <strong>تمامی متدهای abstract</strong> تعریف‌شده در ابرکلاس را <strong>پیاده‌سازی</strong> نکند، خودِ این زیرکلاس هم باید به‌عنوان <strong>abstract</strong> تعریف شود.  
  این قاعده تضمین می‌کند که هر کلاسی که قرار است <strong>قابل نمونه‌سازی</strong> باشد، همهٔ رفتارهای لازم را داشته باشد.  
  در غیر این صورت، کلاس همچنان ناقص است و نباید بتوان از آن شیء ساخت.</p>

  <p>به بیان دیگر، وقتی یک <strong>زیرکلاس غیر-abstract</strong> از یک <strong>کلاس abstract</strong> <strong>ارث‌بری</strong> می‌کند، موظف است <strong>تمامی متدهای abstract</strong> را <strong>پیاده‌سازی</strong> کند، حتی اگر در آن زیرکلاس مستقیماً از همهٔ آن متدها استفاده نشود.  
  این الزام باعث می‌شود که هر کلاس غیر-abstract <strong>قابلیت نمونه‌سازی کامل</strong> داشته باشد و هیچ متد تعریف‌نشده‌ای باقی نماند.</p>

  <p><strong>کلاسی</strong> با <strong>متدهای abstract</strong> باید به‌صورت <strong>abstract</strong> تعریف شود.</p>

  <p>اما می‌توانیم یک <strong>کلاس abstract</strong> را بدون <strong>متدهای abstract</strong> تعریف کنیم.</p>

  <p>قادر نیستید با <strong>new</strong> از یک <strong>کلاس abstract</strong> <strong>نمونه</strong> (شیء) ایجاد کنید.</p>

  <p>از این <strong>کلاس abstract</strong> می‌توان به‌عنوان <strong>کلاس والد</strong> برای تعریف <strong>زیرکلاس‌های جدید</strong> استفاده کرد.</p>


  <hr class="divider" />
</div>

```python
abstract class Shape {
    public abstract double area();

    public void display() {
        System.out.println("This is a shape.");
    }
}

class Circle extends Shape {
    private double radius;

    public Circle(double radius) {
        this.radius = radius;
    }

    @Override
    public double area() {
        return Math.PI * radius * radius;
    }
}

abstract class Polygon extends Shape {
}

class Rectangle extends Polygon {
    private double width;
    private double height;

    public Rectangle(double width, double height) {
        this.width = width;
        this.height = height;
    }

    @Override
    public double area() {
        return width * height;
    }
}

public class Main {
    public static void main(String[] args) {
        Circle c = new Circle(5);
        Rectangle r = new Rectangle(4, 6);

        c.display();
        System.out.println("Circle area: " + c.area());

        r.display();
        System.out.println("Rectangle area: " + r.area());
    }
}
```

  <h3>آیا ابرکلاس یک کلاس انتزاعی می‌تواند خود یک کلاس غیرانتزاعی باشد؟</h3>
  <p>یک <strong>زیرکلاس abstract</strong> می‌تواند <strong>والدی غیر-abstract</strong> داشته باشد.</p>

  <p>برای مثال، کلاس <strong>Object</strong> یک <strong>کلاس عادی</strong> (غیر-abstract) است، اما هر <strong>کلاس abstract</strong> می‌تواند فرزند آن باشد.</p>

  <p>به مثال زیر دقت کنید:</p>

  <hr class="divider" />
</div>

```python
class Vehicle {
    public void start() {
        System.out.println("Vehicle is starting...");
    }
}

abstract class Car extends Vehicle {
    public abstract void drive();
}

class ElectricCar extends Car {
    @Override
    public void drive() {
        System.out.println("ElectricCar is driving silently...");
    }
}

public class Main {
    public static void main(String[] args) {
        ElectricCar tesla = new ElectricCar();
        tesla.start();
        tesla.drive();
    }
}
```

  <h3>توضیح کد</h3>
  <p>کلاس <strong>Vehicle</strong> یک <strong>کلاس غیر-abstract</strong> است که متدی به نام <strong>start</strong> دارد و پیاده‌سازی آن آمادهٔ استفاده است. این کلاس به‌عنوان <strong>والد</strong> برای دیگر کلاس‌ها عمل می‌کند.</p>

  <p>کلاس <strong>Car</strong> یک <strong>زیرکلاس abstract</strong> است که از <strong>Vehicle</strong> ارث‌بری می‌کند. این کلاس یک <strong>متد abstract</strong> به نام <strong>drive</strong> دارد که هیچ پیاده‌سازی مشخصی ندارد و باید در زیرکلاس‌های فرزند پیاده‌سازی شود. از آنجا که Car یک کلاس abstract است، نمی‌توان از آن مستقیماً <strong>نمونه‌سازی</strong> کرد.</p>

  <p>کلاس <strong>ElectricCar</strong> یک <strong>زیرکلاس غیر-abstract</strong> است که از <strong>Car</strong> ارث‌بری می‌کند و متد <strong>drive</strong> را پیاده‌سازی کرده است. از آنجا که این کلاس تمام متدهای abstract والد خود را پیاده‌سازی کرده، می‌توان از آن <strong>نمونه‌سازی</strong> کرد.</p>

  <p>در کلاس <strong>Main</strong> یک شیء از نوع <strong>ElectricCar</strong> ساخته می‌شود. سپس متد <strong>start</strong> که از کلاس <strong>Vehicle</strong> به ارث برده شده و متد <strong>drive</strong> که در کلاس <strong>ElectricCar</strong> پیاده‌سازی شده، فراخوانی می‌شوند. در نتیجه خروجی برنامه ابتدا پیام شروع وسیله و سپس پیام رانندگی خودروی برقی را چاپ می‌کند.</p>

  <p>در این مثال یاد گرفتیم که:</p>
  <ul style="margin-right:20px;">
    <li>یک <strong>زیرکلاس abstract</strong> می‌تواند از یک <strong>کلاس غیر-abstract</strong> ارث‌بری کند.</li>
    <li>تا زمانی که متدهای <strong>abstract</strong> والد در زیرکلاس پیاده‌سازی نشوند، نمی‌توان از آن زیرکلاس <strong>نمونه‌سازی</strong> کرد.</li>
    <li>زمانی که یک زیرکلاس همهٔ متدهای abstract والد را پیاده‌سازی می‌کند، خودش تبدیل به یک کلاس <strong>غیر-abstract</strong> شده و قابلیت نمونه‌سازی پیدا می‌کند.</li>
    <li>متدهای ارث‌بری‌شده از کلاس والد عادی می‌توانند مستقیماً توسط نمونهٔ کلاس فرزند استفاده شوند.</li>
  </ul>

  <h3>یکی از کاربردهای کلاس انتزاعی</h3>
  <p>یکی از مهم‌ترین <strong>کاربردهای کلاس abstract</strong> این است که می‌تواند برای <strong>پیاده‌سازی بخشی از متدهای یک interface</strong> مورد استفاده قرار گیرد. این کار باعث می‌شود که دیگر لازم نباشد هر کلاسی که از آن interface استفاده می‌کند، همهٔ متدهای آن را از ابتدا پیاده‌سازی کند.</p>

<p>در واقع، کلاس abstract به عنوان یک <strong>لایهٔ میانی</strong> بین interface و کلاس‌های معمولی عمل می‌کند. به این ترتیب، بخش‌هایی از پیاده‌سازی که برای همهٔ کلاس‌ها یکسان است، فقط یک بار در کلاس abstract نوشته می‌شود و کلاس‌های فرزند می‌توانند از آن استفاده کنند یا متدهای خاص خود را بازنویسی کنند.</p>

<p><strong>مزایای این رویکرد:</strong></p>
<ul style="margin-left:20px;">
  <li>کاهش <strong>تکرار کد</strong> زیرا پیاده‌سازی متدهای مشترک در یک کلاس abstract انجام می‌شود.</li>
  <li>افزایش <strong>خوانایی و نگه‌داری</strong> کد چون تغییرات در منطق مشترک فقط در یک نقطه (کلاس abstract) انجام می‌شود.</li>
  <li>امکان اعمال <strong>قرارداد (Contract)</strong> از طریق interface و در عین حال دادن پیاده‌سازی پایه از طریق کلاس abstract.</li>
</ul>

<p><strong>نمونهٔ مفهومی:</strong></p>

<p><strong>توضیح مثال:</strong></p>
<ul style="margin-left:20px;">
  <li><strong>interface Drawable</strong> قراردادی تعریف می‌کند که هر کلاسی که آن را پیاده‌سازی کند باید متدهای <strong>draw</strong> و <strong>resize</strong> را داشته باشد.</li>
  <li>کلاس <strong>Shape</strong> یک <strong>کلاس abstract</strong> است که متد <strong>resize</strong> را پیاده‌سازی کرده و متد <strong>draw</strong> را به کلاس‌های فرزند واگذار می‌کند.</li>
  <li>کلاس <strong>Circle</strong> از کلاس <strong>Shape</strong> ارث‌بری می‌کند و متد <strong>draw</strong> را پیاده‌سازی می‌کند.</li>
  <li>در متد <strong>main</strong> می‌توان از قابلیت‌های هم کلاس abstract و هم کلاس فرزند استفاده کرد.</li>
</ul>

<p>این روش به توسعه‌دهندگان اجازه می‌دهد بخشی از پیاده‌سازی را یکجا در کلاس abstract انجام دهند و بقیهٔ جزئیات را به کلاس‌های فرزند بسپارند. این امر انعطاف‌پذیری، کاهش تکرار کد و معماری تمیزتر در پروژه‌های بزرگ را به همراه دارد.</p>

  <hr class="divider" />

```python
interface Drawable {
    void draw();
    void resize();
}

abstract class Shape implements Drawable {
    public void resize() {
        System.out.println("Resizing shape...");
    }
}

class Circle extends Shape {
    public void draw() {
        System.out.println("Drawing circle");
    }
}

class Main {
    public static void main(String[] args) {
        Shape c = new Circle();
        c.draw();
        c.resize();
    }
}
```

  <h3>Interface</h3>
  <p><strong>Interface</strong> در جاوا یک <strong>طرح کلی (Contract)</strong> برای کلاس‌ها است که فقط شامل <strong>تعاریف متدها</strong> و <strong>ثابت‌های استاتیک</strong> می‌باشد و هیچ پیاده‌سازی (بدنه متد) ندارد. تمام متدهای یک interface به‌صورت <strong>abstract</strong> در نظر گرفته می‌شوند.</p>

  <p><strong>Interface</strong> در جاوا سازوکاری برای ایجاد <strong>انتزاع (Abstraction)</strong> کامل است. به این معنی که با استفاده از interface، می‌توان یک لایهٔ کاملاً انتزاعی (%100 abstract) تعریف کرد تا کلاس‌ها از آن پیروی کنند.</p>

  <p>ویژگی‌های کلیدی interface:</p>
  <ul style="margin-right:2em;">
    <li>در interface تمام متدها به صورت <strong>abstract</strong> هستند و پیاده‌سازی ندارند.</li>
    <li>به ما اجازه می‌دهد تا <strong>انتزاع کامل</strong> داشته باشیم و پیاده‌سازی متدها را به کلاس‌هایی که از interface استفاده می‌کنند واگذار کنیم.</li>
    <li>امکان <strong>ارث‌بری چندگانه</strong> را فراهم می‌کند، زیرا یک کلاس می‌تواند چندین interface را پیاده‌سازی کند.</li>
    <li>مانند کلاس‌های <strong>abstract</strong>، نمی‌توانیم از interface مستقیماً نمونه‌سازی کنیم.</li>
  </ul>

  <p>با استفاده از interface می‌توان قراردادهای مشخصی را برای کلاس‌ها تعریف کرد و هر کلاس موظف است متدهای تعریف‌شده در interface را پیاده‌سازی کند. این رویکرد باعث انعطاف‌پذیری و قابلیت توسعه در پروژه‌های بزرگ می‌شود.</p>

  <h3>فهم رابطه بین کلاس‌ها و Interface ها</h3>

  <p><strong>رابطه <strong>کلاس</strong> با <strong>کلاس</strong>:</strong> در <strong>جاوا</strong> یک <strong>کلاس</strong> می‌تواند از یک <strong>کلاس</strong> دیگر <strong>ارث‌بری</strong> کند. این رابطه به کمک کلیدواژه <strong>extends</strong> تعریف می‌شود. در این حالت، <strong>کلاس فرزند</strong> ویژگی‌ها و متدهای قابل ارث را از <strong>کلاس والد</strong> دریافت می‌کند.</p>

  <p><strong>رابطه <strong>کلاس</strong> با <strong>اینترفیس</strong>:</strong> یک <strong>کلاس</strong> برای <strong>پیاده‌سازی</strong> قرارداد یک <strong>اینترفیس</strong> از کلیدواژه <strong>implements</strong> استفاده می‌کند. یعنی <strong>کلاس</strong> متعهد می‌شود که تمام <strong>متدهای abstract</strong> تعریف‌شده در آن <strong>اینترفیس</strong> را پیاده‌سازی کند.</p>

  <p><strong>رابطه <strong>اینترفیس</strong> با <strong>اینترفیس</strong>:</strong> یک <strong>اینترفیس</strong> می‌تواند از یک یا چند <strong>اینترفیس</strong> دیگر <strong>ارث‌بری</strong> کند. در این حالت از کلیدواژه <strong>extends</strong> استفاده می‌شود. این کار برای ترکیب چند <strong>قرارداد</strong> در یک <strong>اینترفیس</strong> جدید مفید است.</p>

  <p><strong>نکات :</strong></p>
  <ul style="margin-right:2em">
    <li><strong>کلاس‌ها</strong> فقط می‌توانند از یک <strong>کلاس</strong> دیگر <strong>ارث‌بری</strong> کنند.</li>
    <li><strong>کلاس‌ها</strong> می‌توانند چندین <strong>اینترفیس</strong> را <strong>پیاده‌سازی</strong> کنند.</li>
    <li>یک <strong>اینترفیس</strong> می‌تواند از چندین <strong>اینترفیس</strong> دیگر <strong>ارث‌بری</strong> کند.</li>
    <li>هیچ <strong>اینترفیس</strong>ی از یک <strong>کلاس</strong> <strong>ارث‌بری</strong> نمی‌کند.</li>
    <li>هیچ <strong>کلاس</strong>ی نمی‌تواند با <strong>extends</strong> از یک <strong>اینترفیس</strong> <strong>ارث‌بری</strong> کند.</li>
  </ul>

  <p><strong>ارث‌بری</strong> در <strong>کلاس‌ها</strong> برای <strong>به‌اشتراک‌گذاری پیاده‌سازی</strong> و <strong>ویژگی‌ها</strong> استفاده می‌شود، در حالی که استفاده از <strong>اینترفیس</strong> برای تعریف <strong>قراردادهای رفتاری</strong> و <strong>توانایی‌های مشترک</strong> بین <strong>کلاس‌های نامرتبط</strong> کاربرد دارد.</p>

  <p><strong>ماهیت interface:</strong> در جاوا، <strong>interface</strong> مانند یک <strong>کلاس ویژه</strong> است که از آن برای تعریف <strong>قراردادهای رفتاری</strong> استفاده می‌شود. با این حال، برخلاف کلاس معمولی، نمی‌توان از آن نمونه‌سازی (instantiate) کرد.</p>

<p><strong>نیاز به فایل جداگانه:</strong> هر <strong>interface</strong> باید در یک <strong>فایل جداگانه</strong> تعریف شود. این فایل با همان نام interface ذخیره می‌شود و ساختار آن مشابه فایل کلاس است، اما به جای کلیدواژه <strong>class</strong> از <strong>interface</strong> استفاده می‌شود.</p>

<p><strong>عدم امکان نمونه‌سازی مستقیم:</strong> مشابه <strong>کلاس abstract</strong>، نمی‌توان با استفاده از کلیدواژه <strong>new</strong> از یک <strong>interface</strong> شیء ایجاد کرد. زیرا interface فقط تعریف‌کننده <strong>رفتار</strong> است و هیچ پیاده‌سازی در خود ندارد.</p>

<p><strong>کاربرد در الگوهای مشابه:</strong> در بسیاری از موارد، از <strong>interface</strong> برای الگوهای مشابهی که نیاز به تعیین <strong>رفتار مشترک</strong> بین کلاس‌های متفاوت دارند استفاده می‌شود. این رفتار می‌تواند مستقل از ساختار و سلسله‌مراتب کلاس‌ها باشد.</p>

<p><strong>استفاده به‌عنوان نوع داده:</strong> می‌توان از یک <strong>interface</strong> به‌عنوان <strong>نوع داده</strong> برای یک <strong>متغیر</strong> استفاده کرد. به این ترتیب، آن متغیر می‌تواند هر شیئی را که آن interface را پیاده‌سازی کرده است، در خود نگه دارد.</p>

  <h3>ارث‌بری چندگانه در جاوا به کمک Interface</h3>


  <div class="images">
    <img src="img/image9.png" alt="Compiler Picture">
  </div>
  <hr class="divider" />
  <p>
  <strong>ارث‌بری چندگانه در جاوا (Multiple Inheritance in Java)</strong> مفهومی است که به یک کلاس یا یک اینترفیس اجازه می‌دهد از چندین اینترفیس ارث‌بری کند یا آن‌ها را پیاده‌سازی نماید.
  </p>

  <p>
  در جاوا، <strong>کلاس‌ها</strong> نمی‌توانند از چند کلاس دیگر ارث‌بری کنند زیرا این کار باعث ایجاد <strong>ابهام</strong> (مانند مسئله الماس) می‌شود. اما جاوا این امکان را با استفاده از <strong>اینترفیس‌ها</strong> فراهم کرده است تا یک کلاس بتواند چندین اینترفیس را <strong>پیاده‌سازی</strong> کند یا یک اینترفیس از چندین اینترفیس دیگر <strong>ارث‌بری</strong> نماید.
  </p>

  <p>
  <strong>تعریف:</strong> اگر یک <strong>کلاس</strong> بیش از یک <strong>اینترفیس</strong> را پیاده‌سازی کند یا یک <strong>اینترفیس</strong> از چندین <strong>اینترفیس</strong> دیگر ارث‌بری کند، به این ویژگی <strong>ارث‌بری چندگانه</strong> گفته می‌شود.
  </p>

  <p>
  در شکل نمایش داده شده:
  </p>

  <p style="margin-right: 2em;">
  - در سمت چپ، یک <strong>کلاس</strong> وجود دارد که چندین <strong>اینترفیس</strong> را <strong>پیاده‌سازی</strong> کرده است. این نشان می‌دهد که یک کلاس می‌تواند رفتارهای تعریف شده در چندین اینترفیس را در خود پیاده‌سازی نماید.
  </p>

  <p style="margin-right: 2em;">
  - در سمت راست، یک <strong>اینترفیس</strong> از چندین <strong>اینترفیس دیگر</strong> <strong>ارث‌بری</strong> می‌کند. به این ترتیب، این اینترفیس جدید تمام متدهای انتزاعی (abstract methods) تعریف شده در اینترفیس‌های والد را نیز به ارث می‌برد.
  </p>

  <p>
  <strong>مزیت اصلی:</strong> این روش اجازه می‌دهد بدون مشکل ابهام در ارث‌بری، یک کلاس یا اینترفیس چندین رفتار مختلف را از منابع گوناگون به ارث ببرد یا پیاده‌سازی کند.
  </p>

  <p>
  <strong>نکته:</strong> این رویکرد مخصوصاً برای مدیریت <strong>چندین قابلیت</strong> (مثل قابل‌خوردن بودن، قابل‌چاپ بودن، و غیره) در یک کلاس مفید است، بدون اینکه نیاز باشد آن قابلیت‌ها را از کلاس‌های مختلف به ارث ببریم.
  </p>

  <hr class="divider" />

</div>




```python
interface Printable {
    void print();
}

interface Showable {
    void show();
}

class A7 implements Printable, Showable {
    public void print() {
        System.out.println("Hello");
    }
    public void show() {
        System.out.println("Welcome");
    }

    public static void main(String args[]) {
        A7 obj = new A7();
        obj.print();
        obj.show();
    }
}
```

  <h3>ارث‌بری چندگانه در جاوا به کمک Interface</h3>


<p>
در زبان <strong>جاوا</strong>، <strong>ارث‌بری چندگانه</strong> (Multiple Inheritance) برای <strong>کلاس‌ها</strong> پشتیبانی نمی‌شود. به این معنی که یک کلاس نمی‌تواند به‌صورت مستقیم از چند کلاس مختلف ارث‌بری کند. دلیل این محدودیت جلوگیری از بروز <strong>ابهام</strong> در ساختار برنامه و پیچیدگی‌های احتمالی است.
</p>

<p>
اما با استفاده از <strong>interface</strong> می‌توان این محدودیت را برطرف کرد. زمانی که یک کلاس چندین <strong>interface</strong> را <strong>پیاده‌سازی</strong> (implements) می‌کند، هیچ ابهامی ایجاد نمی‌شود زیرا <strong>interface</strong>ها تنها شامل <strong>تعریف متدها</strong> هستند و پیاده‌سازی ندارند.
</p>

<p>
بنابراین، با کمک <strong>interface</strong> می‌توان به نوعی <strong>ارث‌بری چندگانه</strong> دست یافت، بدون اینکه مشکلات ناشی از ارث‌بری چندگانه کلاس‌ها به‌وجود آید. این ویژگی یکی از مهم‌ترین کاربردهای <strong>interface</strong> در جاوا محسوب می‌شود.
</p>


  <hr class="divider" />

</div>




```python
interface Printable{  
void print();  
}  

interface Showable{  
void print();  
}  

class testinterface1 implements Printable,Showable{  

public void print(){System.out.println("Hello");}  

public static void main(String args[]){  
testinterface1 obj = new testinterface1();  
obj.print();  
 }  
}  
```

```python
interface Printable {
    void print();
}

interface Showable extends Printable {
    void show();
}

class Testinterface2 implements Showable {

    public void print() {
        System.out.println("Hello");
    }

    public void show() {
        System.out.println("Welcome");
    }

    public static void main(String args[]) {
        Testinterface2 obj = new Testinterface2();
        obj.print();
        obj.show();
    }
}
```

```python
interface A {
    void a();
    void b();
    void c();
    void d();
}

abstract class B implements A {
    public void c() {
        System.out.println("I am C");
    }
}

class M extends B {
    public void a() {
        System.out.println("I am a");
    }
    public void b() {
        System.out.println("I am b");
    }
    public void d() {
        System.out.println("I am d");
    }
}

class Test5 {
    public static void main(String args[]) {
        A a = new M();
        a.a();
        a.b();
        a.c();
        a.d();
    }
}

/*
output:
    I am a  
    I am b  
    I am c  
    I am d  
*/
```

<p>در جاوا، زمانی که یک <strong>interface</strong> هیچ متد یا فیلدی ندارد، به آن <strong>Marker Interface</strong> یا <strong>Tagged Interface</strong> گفته می‌شود. این نوع اینترفیس‌ها تنها برای انتقال اطلاعات متادیتا به <strong>JVM</strong> استفاده می‌شوند و هیچ رفتار خاصی را تعریف نمی‌کنند.</p>

<p>برخی از مثال‌های معروف از این نوع اینترفیس‌ها عبارت‌اند از: <strong>Serializable</strong>، <strong>Cloneable</strong> و <strong>Remote</strong>. این اینترفیس‌ها برای مشخص کردن یک ویژگی خاص در کلاس‌ها استفاده می‌شوند؛ مثلاً وقتی یک کلاس <strong>Serializable</strong> را پیاده‌سازی می‌کند، در واقع به <strong>JVM</strong> اطلاع می‌دهد که این شیء قابل ذخیره‌سازی یا انتقال است.</p>

<p>در زمان اجرا، <strong>JVM</strong> با بررسی اینکه آیا یک کلاس، یک <strong>Marker Interface</strong> را پیاده‌سازی کرده است یا خیر، تصمیم می‌گیرد عملیات خاصی را انجام دهد. برای نمونه، اگر کلاسی <strong>Serializable</strong> باشد، <strong>JVM</strong> اجازه‌ی نوشتن و خواندن آن شیء در جریان داده را می‌دهد.</p>

  <h3>برخی از واسط‌های محبوب در جاوا</h3>

<p><strong>Iterator</strong>: برای پیمایش بر روی مجموعه‌ای (<strong>collection</strong>) از اشیاء بدون نیاز به دانستن نحوه‌ی ذخیره‌سازی آن‌ها استفاده می‌شود. برای مثال در <strong>list</strong> یا <strong>set</strong> از این رابط استفاده می‌شود تا بتوان به ترتیب به عناصر دسترسی پیدا کرد. توضیحات کامل‌تر آن در بخش <strong>Collections</strong> ارائه می‌شود.</p>

<p><strong>Cloneable</strong>: برای ایجاد یک کپی از شیء موجود از طریق متد <strong>()clone</strong> که در کلاس <strong>Object</strong> قرار دارد استفاده می‌شود. اگر یک کلاس این اینترفیس را پیاده‌سازی کند، نشان می‌دهد که اجازه‌ی کپی‌سازی ایمن از آن شیء وجود دارد.</p>

<p><strong>Serializable</strong>: برای <strong>بسته‌بندی (pack)</strong> و ذخیره‌سازی اشیاء به گونه‌ای که بتوان آن‌ها را از طریق شبکه انتقال داد یا بر روی دیسک ذخیره کرد، استفاده می‌شود. هدف این است که شیء به بایت‌ها تبدیل شده و سپس قابل بازیابی (بازسازی) باشد. <strong>JVM</strong> با استفاده از این اینترفیس می‌تواند عملیات مربوط به ورودی/خروجی (<strong>I/O</strong>) را بر روی اشیاء انجام دهد.</p>

<p><strong>Comparable</strong>: برای ایجاد یک ترتیب کلی (<strong>total order</strong>) بین اشیاء به کار می‌رود. این اینترفیس معمولاً برای مرتب‌سازی مجموعه‌ها یا مقایسه‌ی دو شیء با هم استفاده می‌شود.</p>

  <h3>واسط Iterator</h3>

<p><strong>Iterator</strong> یکی از رابط‌های مهم در پکیج <strong>java.util</strong> است که برای پیمایش بر روی عناصر ذخیره‌شده در مجموعه‌ها (<strong>collections</strong>) مورد استفاده قرار می‌گیرد. این واسط روشی استاندارد برای دسترسی ترتیبی به اشیاء بدون نیاز به دانستن ساختار داخلی مجموعه فراهم می‌کند. با استفاده از <strong>Iterator</strong> می‌توان به ترتیب عناصر را بررسی، خوانده و حتی حذف کرد.</p>

<hr class="divider" />
<p><strong>متد ()hasNext</strong> بررسی می‌کند که آیا در مجموعه (collection) عنصر دیگری برای پیمایش وجود دارد یا خیر. اگر عنصر بعدی وجود داشته باشد مقدار <strong>true</strong> برمی‌گرداند و در غیر این صورت مقدار <strong>false</strong>. این متد برای کنترل ادامه‌ی حلقه در هنگام پیمایش مجموعه استفاده می‌شود.</p>

<p><strong>متد ()next</strong> عنصر بعدی موجود در مجموعه را بازمی‌گرداند. در صورتی که دیگر عنصری برای بازگشت وجود نداشته باشد، این متد استثنای <strong>NoSuchElementException</strong> را پرتاب می‌کند. معمولاً این متد همراه با <strong>()hasNext</strong> به کار می‌رود تا در زمان پیمایش، از وجود عنصر بعدی اطمینان حاصل شود.</p>

<p><strong>متد ()remove</strong> برای حذف عنصر فعلی از مجموعه استفاده می‌شود. این متد تنها در صورتی مجاز به اجراست که پیش از آن متد <strong>()next</strong> فراخوانی شده باشد. در غیر این صورت، اجرای آن باعث پرتاب استثنای <strong>IllegalStateException</strong> می‌شود. استفاده از این متد به شما امکان می‌دهد تا در حین پیمایش، عناصر خاصی را از مجموعه حذف کنید بدون آنکه به ساختار داخلی مجموعه دست بزنید.</p>

  <h3>واسط Clonable</h3>

<p><strong>Cloneable</strong> یک واسط نشانه‌گذاری در جاوا است که برای مشخص کردن قابلیت <strong>کپی‌سازی (cloning)</strong> در کلاس‌ها استفاده می‌شود. این واسط به <strong>JVM</strong> اطلاع می‌دهد که اشیای ساخته‌شده از آن کلاس می‌توانند با اطمینان کپی شوند.</p>

<p>این واسط هیچ متد یا فیلدی ندارد و تنها به‌عنوان یک علامت برای <strong>JVM</strong> عمل می‌کند. اگر کلاسی این واسط را پیاده‌سازی نکرده باشد و متد <strong>()clone</strong> روی شیء آن فراخوانی شود، برنامه با استثنای <strong>CloneNotSupportedException</strong> مواجه خواهد شد.</p>

<p>در مثال کلاس <strong>Car</strong>، واسط <strong>Cloneable</strong> پیاده‌سازی شده تا اشیای این کلاس قابل کپی باشند. متد <strong>()clone</strong> در این کلاس با ایجاد یک شیء جدید از نوع <strong>Car</strong> و کپی کردن مقادیر ویژگی‌های <strong>make</strong>، <strong>model</strong> و <strong>price</strong>، نسخه‌ای جدید از شیء فعلی تولید می‌کند.</p>

// Car example revisited
public class Car implements Cloneable {
    private String make;
    private String model;
    private double price;

    // default constructor
    public Car() {
        this("", "", 0.0);
    }

    // give reasonable values to instance variables
    public Car(String make, String model, double price) {
        this.make = make;
        this.model = model;
        this.price = price;
    }

    // the Cloneable interface
    public Object clone() {
        return new Car(this.make, this.model, this.price);
    }
}
